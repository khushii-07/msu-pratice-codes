```python
# proj01.py
# CSE 231 - Project 1 (Spring 2010)
# This program calculates time dilation and mass increase at near-light speeds.

import math

SPEED_OF_LIGHT = 299_792_458
REST_MASS_KG = 70000

percent_light_speed = float(input("Enter the speed as a percentage of the speed of light (e.g. 90 for 90%): "))

v = (percent_light_speed / 100) * SPEED_OF_LIGHT
factor = 1 / math.sqrt(1 - (v ** 2) / (SPEED_OF_LIGHT ** 2))
adjusted_mass = REST_MASS_KG * factor

distances = {
    "Alpha Centauri": 4.3,
    "Barnard's Star": 6.0,
    "Betelgeuse": 309,
    "Andromeda Galaxy": 2_000_000
}

print("\nResults:")
print("Speed entered: {:.2f}% of the speed of light".format(percent_light_speed))
print("Relativistic mass: {:.2f} kg".format(adjusted_mass))
print("Time experienced by astronauts:")

for star, distance in distances.items():
    time = distance / factor
    print("- {}: {:.2f} years".format(star, time))
```

```python
from math import pow

print("Music Note to Frequency Converter")
print("Enter 3 notes (octave and pitch class)")

oct1 = int(input("Octave for note 1: "))
pc1 = int(input("Pitch class for note 1 (0-11): "))
oct2 = int(input("Octave for note 2: "))
pc2 = int(input("Pitch class for note 2 (0-11): "))
oct3 = int(input("Octave for note 3: "))
pc3 = int(input("Pitch class for note 3 (0-11): "))

ref_oct = 4
ref_pc = 9
ref_hz = 440

def get_freq(octave, pitch):
    o = octave - ref_oct
    m = pitch - ref_pc
    return pow(2, o + m / 12.0) * ref_hz

hz1 = get_freq(oct1, pc1)
hz2 = get_freq(oct2, pc2)
hz3 = get_freq(oct3, pc3)

print("\nFrequencies:")
print("Note 1:", oct1, ".", pc1, "=", round(hz1, 2), "Hz")
print("Note 2:", oct2, ".", pc2, "=", round(hz2, 2), "Hz")
print("Note 3:", oct3, ".", pc3, "=", round(hz3, 2), "Hz")
```

```python
def fact(n):
    f = 1
    for i in range(2, n+1):
        f *= i
    return f

def sin(x):
    s = 0
    for k in range(10):
        s += ((-1)**k * x**(2*k+1)) / fact(2*k+1)
    return s

def cos(x):
    c = 0
    for k in range(10):
        c += ((-1)**k * x**(2*k)) / fact(2*k)
    return c

while True:
    ch = input("A:Fact B:Sin C:Cos Q:Quit: ").lower()
    if ch == 'a':
        print("Factorial:", fact(int(input("Enter number: "))))
    elif ch == 'b':
        print("Sine:", sin(float(input("Enter x in radians: "))))
    elif ch == 'c':
        print("Cosine:", cos(float(input("Enter x in radians: "))))
    elif ch == 'q':
        break
```

```python
def additive(n):
    count = 0
    while n >= 10:
        n = sum(int(d) for d in str(n))
        count += 1
    return count, n

def multiplicative(n):
    count = 0
    while n >= 10:
        prod = 1
        for d in str(n):
            prod *= int(d)
        n = prod
        count += 1
    return count, n

while True:
    num = int(input("Enter a number (negative to quit): "))
    if num < 0:
        break
    if num < 10:
        print(f"Additive persistence: 0, root: {num}")
        print(f"Multiplicative persistence: 0, root: {num}")
    else:
        a_count, a_root = additive(num)
        m_count, m_root = multiplicative(num)
        print(f"Additive persistence: {a_count}, root: {a_root}")
        print(f"Multiplicative persistence: {m_count}, root: {m_root}")
```

```python
def get_number(line):
    return line.split()[0]

def get_balance(parts):
    return float(parts[1])

def get_name(parts):
    return " ".join(parts[2:])

def equal_floats(x, y):
    return abs(x - y) < 1.0e-8

prefix = input("Enter file name prefix: ")

try:
    old = open(prefix + ".old.txt", "r")
    new = open(prefix + ".new.txt", "w")
except:
    print("File error")
    exit()

for line in old:
    line = line.strip()
    if line == "999999":
        new.write("999999\n")
        break

    parts = line.split()
    if len(parts) < 3:
        print("Invalid line format.")
        continue

    acc = get_number(line)
    try:
        bal = get_balance(parts)
    except:
        print("Invalid balance format.")
        continue
    name = get_name(parts)

    print("Account:", acc, "Balance:", bal, "Name:", name)

    while True:
        code = input("Transaction (d/w/c/a): ").lower()
        if code == "a":
            new.write(f"{acc} {bal:10.2f} {name}\n")
            break
        elif code == "d":
            try:
                amt = float(input("Deposit amount: "))
                if bal + amt <= 9999999.99:
                    bal += amt
                else:
                    print("Amount too large.")
            except:
                print("Invalid deposit amount.")
        elif code == "w":
            try:
                amt = float(input("Withdraw amount: "))
                if bal - amt >= 0:
                    bal -= amt
                else:
                    print("Insufficient balance.")
            except:
                print("Invalid withdrawal amount.")
        elif code == "c":
            if equal_floats(bal, 0.0):
                print("Account closed.")
                break
            else:
                print("Balance not zero. Cannot close account.")
        else:
            print("Invalid transaction code.")

old.close()
new.close()
```

```python
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

def reverse_number(n):
    return int(str(n)[::-1])

def main():
    start = int(input("Enter the start of the range: "))
    end = int(input("Enter the end of the range: "))

    natural_palindromes = 0
    non_lychrel = 0
    lychrel_count = 0

    print("\nChecking numbers...\n")

    for num in range(start, end + 1):
        if is_palindrome(num):
            natural_palindromes += 1
            continue

        current = num
        found_palindrome = False

        for _ in range(60):
            reversed_num = reverse_number(current)
            current += reversed_num

            if is_palindrome(current):
                non_lychrel += 1
                found_palindrome = True
                break

        if not found_palindrome:
            lychrel_count += 1
            print(f"Lychrel number found: {num}")

    print("\nStatistics:")
    print(f"Natural palindromes: {natural_palindromes}")
    print(f"Non-Lychrel numbers: {non_lychrel}")
    print(f"Lychrel numbers: {lychrel_count}")

if __name__ == "__main__":
    main()
```

```python
def is_valid(num):
    return num.isdigit() and len(num) == 5 and len(set(num)) == 5

def get_secret():
    while True:
        secret = input("Enter a 5-digit number (no repeated digits): ")
        if is_valid(secret):
            print("\n" * 20)
            return secret
        else:
            print("Invalid number. Try again.")

def get_guess():
    while True:
        guess = input("Enter your guess (or type QUIT): ")
        if guess.upper() == "QUIT":
            return "QUIT"
        if is_valid(guess):
            return guess
        else:
            print("Invalid guess. Try again.")

def count_matches(secret, guess):
    correct_digits = 0
    correct_places = 0
    for i in range(5):
        if guess[i] in secret:
            correct_digits += 1
        if guess[i] == secret[i]:
            correct_places += 1
    return correct_digits, correct_places

def main():
    print("Welcome to the Guessing Game!")
    print("You have 10 tries to guess the secret number.")

    secret = get_secret()
    tries = 0
    max_tries = 10

    while tries < max_tries:
        guess = get_guess()
        if guess == "QUIT":
            print("You quit the game.")
            break

        tries += 1
        correct_digits, correct_places = count_matches(secret, guess)

        print(f"Guess #{tries}")
        print("Correct digits:", correct_digits)
        print("Correct places:", correct_places)

        if guess == secret:
            print("You guessed it in", tries, "tries!")
            break
    else:
        print("Out of tries! The number was", secret)

main()
```

```python
def is_valid_letter(ch):
    return len(ch) == 1 and ch.isalpha()

def display_phrase(secret, guesses):
    return ''.join([ch if ch.lower() in guesses or not ch.isalpha() else '_' for ch in secret])

def main():
    print("Welcome to Hangman!")
    print("You have 6 chances to guess the word or phrase.")
    secret = input("Enter the phrase to guess: ").lower()

    guessed_letters = ""
    wrong_letters = ""
    chances = 6

    while chances > 0:
        display = display_phrase(secret, guessed_letters)

        if "_" not in display:
            print(f"Phrase: {display} | Wrong: {wrong_letters} | Left: {chances}")
            print("You guessed the phrase! You win.")
            break

        print(f"Phrase: {display} | Wrong: {wrong_letters} | Left: {chances}")
        guess = input("Guess a letter: ").lower()

        if not is_valid_letter(guess):
            print("Please enter a valid alphabetic letter.")
            continue

        if guess in guessed_letters or guess in wrong_letters:
            print("You already guessed that letter.")
            continue

        if guess in secret:
            guessed_letters += guess
            print("Correct guess.")
        else:
            wrong_letters += guess
            chances -= 1
            print("Wrong guess.")

    else:
        print("Out of guesses. You lost.")
        print("The phrase was:", secret)

main()
```

```python
def is_valid_guess(guess):
    return guess.isdigit() and len(guess) == 4 and len(set(guess)) == 4

def count_feedback(key, guess):
    position = sum(key[i] == guess[i] for i in range(4))
    exist = sum((d in key) and (key[i] != guess[i]) for i, d in enumerate(guess))
    return exist, position

def main():
    print("Welcome to Mastermind!")
    key = input("Enter a 4-digit key (no repeats): ")
    guesses = []
    attempts = 0
    max_attempts = 12

    while attempts < max_attempts:
        guess = input("Guess: ")

        if len(guess) != 4:
            print("****guess must have length of 4, try again")
            continue
        if not guess.isdigit():
            print("****contains non-numbers, try again")
            continue
        if len(set(guess)) != 4:
            print("****no repeated digits, try again")
            continue

        guesses.append(guess)
        attempts += 1

        if guess == key:
            for g in guesses:
                exist, pos = count_feedback(key, g)
                print(f"{g}: exist:{exist}, position:{pos}")
            print(f"\nYou guessed the key: {key}")
            print(f"It took you {attempts} guesses")
            return

        for g in guesses:
            exist, pos = count_feedback(key, g)
            print(f"{g}: exist:{exist}, position:{pos}")
        print()

    print(f"Sorry, you didn't guess the key in {max_attempts} tries.")
    print(f"The key was: {key}")

main()
```

```python
def rotate_alphabet(n):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    return alphabet[n:] + alphabet[:n]

def encode_message():
    message = input("Enter a message to encode (lowercase only): ")
    rotation = int(input("Enter a rotation (1‚Äì25): "))
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    rotated = rotate_alphabet(rotation)
    result = ""

    for ch in message:
        if ch in alphabet:
            idx = alphabet.index(ch)
            result += rotated[idx]
        else:
            result += ch
    print("Encoded message:", result)

def decode_message():
    message = input("Enter the message to decode: ")
    known_word = input("Enter a word you expect to find in the decoded text: ")
    alphabet = "abcdefghijklmnopqrstuvwxyz"

    for rot in range(1, 26):
        rotated = rotate_alphabet(rot)
        result = ""
        for ch in message:
            if ch in alphabet:
                idx = rotated.index(ch)
                result += alphabet[idx]
            else:
                result += ch
        if known_word in result:
            print("Found rotation:", rot)
            print("Decoded message:", result)
            return
    print("No rotation found with that word.")

def main():
    while True:
        print("\nChoose an option:")
        print("e - encode")
        print("d - decode")
        print("q - quit")

        choice = input("Enter your choice: ").lower()
        if choice == 'e':
            encode_message()
        elif choice == 'd':
            decode_message()
        elif choice == 'q':
            print("Goodbye.")
            break
        else:
            print("Invalid option. Try again.")

main()
```

```python
def roman_to_int(roman):
    roman_values = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100}
    total = 0
    i = 0
    while i < len(roman):
        if i + 1 < len(roman) and roman_values[roman[i]] < roman_values[roman[i + 1]]:
            total += roman_values[roman[i + 1]] - roman_values[roman[i]]
            i += 2
        else:
            total += roman_values[roman[i]]
            i += 1
    return total

def int_to_roman(number):
    result = ""
    values = [
        (100, 'C'), (90, 'XC'), (50, 'L'), (40, 'XL'),
        (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')
    ]
    for val, sym in values:
        while number >= val:
            result += sym
            number -= val
    return result

def main():
    first = input("Enter First Roman Number (no spaces): ").upper()
    first_val = roman_to_int(first)
    print("Value of", first, ":", first_val)

    second = input("Enter Second Roman Number (no spaces): ").upper()
    second_val = roman_to_int(second)
    print("Value of", second, ":", second_val)

    total = first_val + second_val
    print("Digital sum is:", total)

    roman_sum = int_to_roman(total)
    spaced = ' '.join(roman_sum)
    print("Roman sum is:", spaced)

main()
```

```python
import random

def get_computer_choice(rock_count, paper_count, scissor_count):
    if rock_count > paper_count and rock_count > scissor_count:
        return "P"
    elif paper_count > rock_count and paper_count > scissor_count:
        return "S"
    elif scissor_count > rock_count and scissor_count > paper_count:
        return "R"
    else:
        return random.choice(["R", "P", "S"])

def get_result(user, computer):
    if user == computer:
        return "Tie"
    elif (user == "R" and computer == "S") or \
         (user == "P" and computer == "R") or \
         (user == "S" and computer == "P"):
        return "User"
    else:
        return "Computer"

def main():
    print("Welcome to Rock, Paper, Scissors!")

    rock_count = paper_count = scissor_count = 0
    user_wins = computer_wins = ties = 0

    while True:
        choice = input("Enter R, P, S to play or Q to quit: ").strip().upper()

        if choice == 'Q':
            break

        if choice not in ['R', 'P', 'S']:
            print("Invalid choice. Try again.")
            continue

        if choice == 'R':
            rock_count += 1
        elif choice == 'P':
            paper_count += 1
        else:
            scissor_count += 1

        computer = get_computer_choice(rock_count, paper_count, scissor_count)
        print("Computer chose:", computer)

        result = get_result(choice, computer)
        if result == "User":
            print("You win this round!")
            user_wins += 1
        elif result == "Computer":
            print("Computer wins this round!")
            computer_wins += 1
        else:
            print("This round is a tie!")
            ties += 1

    print("\nGame Over")
    print("User wins:", user_wins)
    print("Computer wins:", computer_wins)
    print("Ties:", ties)
    print("Your choices - Rock:", rock_count, "Paper:", paper_count, "Scissors:", scissor_count)

main()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math

LEGAL_COLORS = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'pink']

def get_color_choice(prompt):
    while True:
        color = input(prompt).strip().lower()
        if color in LEGAL_COLORS:
            return color
        else:
            print(f"'{color}' is not a legal choice. Please try again.")

def get_num_hexagons():
    while True:
        try:
            n = int(input("Please enter a number of hexagons per row (4‚Äì20): "))
            if 4 <= n <= 20:
                return n
            else:
                print("It should be between 4 and 20. Please try again.")
        except ValueError:
            print("Invalid input. Please enter an integer between 4 and 20.")

def draw_hexagon(ax, x, y, side_length, color):
    angle_deg = 60
    angle_rad = math.radians(angle_deg)
    hexagon = patches.RegularPolygon((x, y), numVertices=6, radius=side_length / math.cos(math.pi / 6),
                                     orientation=math.radians(30), color=color, ec='black')
    ax.add_patch(hexagon)

def draw_tessellation(n, color1, color2):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal')
    plt.axis('off')

    width = 500
    side_length = width / n / math.sqrt(3)

    dx = side_length * 3 ** 0.5
    dy = side_length * 1.5

    for row in range(n):
        for col in range(n):
            x = col * dx + (row % 2) * (dx / 2)
            y = -row * dy
            color = color1 if (col % 2 == 0) else color2
            draw_hexagon(ax, x, y, side_length, color)

    ax.set_xlim(-side_length, dx * n)
    ax.set_ylim(-dy * n, side_length)
    plt.show()

print("Choices for colors to use are:", ", ".join(LEGAL_COLORS))
color1 = get_color_choice("Please enter your first color choice: ")
color2 = get_color_choice("Please enter your second color choice: ")
n = get_num_hexagons()
draw_tessellation(n, color1, color2)
```

```python
ASCII_LOWERCASE = "abcdefghijklmnopqrstuvwxyz"
ASCII_UPPERCASE = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
DECIMAL_DIGITS  = "0123456789"

def is_alpha(s):
    for c in s:
        if c not in ASCII_LOWERCASE and c not in ASCII_UPPERCASE:
            return False
    return True

def is_digit(s):
    for c in s:
        if c not in DECIMAL_DIGITS:
            return False
    return True

def to_lower(s):
    result = ''
    for c in s:
        if c in ASCII_UPPERCASE:
            index = ASCII_UPPERCASE.index(c)
            result += ASCII_LOWERCASE[index]
        else:
            result += c
    return result

def to_upper(s):
    result = ''
    for c in s:
        if c in ASCII_LOWERCASE:
            index = ASCII_LOWERCASE.index(c)
            result += ASCII_UPPERCASE[index]
        else:
            result += c
    return result

def find_chr(s, ch):
    if len(ch) != 1:
        return -1
    for i in range(len(s)):
        if s[i] == ch:
            return i
    return -1

def find_str(s, substr):
    for i in range(len(s) - len(substr) + 1):
        if s[i:i+len(substr)] == substr:
            return i
    return -1

def replace_chr(s, old, new):
    if len(old) != 1 or len(new) != 1:
        return ''
    result = ''
    for c in s:
        if c == old:
            result += new
        else:
            result += c
    return result

def replace_str(s, old, new):
    if old == '':
        result = new
        for c in s:
            result += c + new
        return result
    result = ''
    i = 0
    while i < len(s):
        if s[i:i+len(old)] == old:
            result += new
            i += len(old)
        else:
            result += s[i]
            i += 1
    return result

# Testing Section
print("Testing Constants")
print(ASCII_LOWERCASE == "abcdefghijklmnopqrstuvwxyz")
print(ASCII_UPPERCASE == "ABCDEFGHIJKLMNOPQRSTUVWXYZ")
print(DECIMAL_DIGITS  == "0123456789")

print("\nTesting is_alpha:")
print(is_alpha("Hello") == True)
print(is_alpha("Hello123") == False)

print("\nTesting is_digit:")
print(is_digit("123456") == True)
print(is_digit("12a34") == False)

print("\nTesting to_lower:")
print(to_lower("HELLO") == "hello")
print(to_lower("Hello123") == "hello123")

print("\nTesting to_upper:")
print(to_upper("hello") == "HELLO")
print(to_upper("Hello123") == "HELLO123")

print("\nTesting find_chr:")
print(find_chr("banana", "a") == 1)
print(find_chr("banana", "x") == -1)
print(find_chr("banana", "ab") == -1)

print("\nTesting find_str:")
print(find_str("banana", "na") == 2)
print(find_str("banana", "xyz") == -1)

print("\nTesting replace_chr:")
print(replace_chr("banana", "a", "o") == "bonono")
print(replace_chr("banana", "na", "oo") == "")

print("\nTesting replace_str:")
print(replace_str("banana", "na", "xy") == "baxyxy")
print(replace_str("banana", "", "*") == "*b*a*n*a*n*a*")
print(replace_str("banana", "x", "y") == "banana")
```

```python
import pandas as pd

def read_gdp(filename, year):
    try:
        with open(filename, 'r') as file:
            data = file.read().split()
            start_year = 1948
            index = (year - start_year) * 4
            gdp_values = list(map(float, data[index:index+4]))
            return sum(gdp_values) / len(gdp_values)
    except Exception as e:
        print(f"Error reading GDP data: {e}")
        return None

def read_unemployment(filename, year):
    try:
        df = pd.read_csv(filename, header=None)
        df.columns = ['Year', 'Month', 'Unemployment']
        year_data = df[df['Year'] == year]['Unemployment']
        return year_data.mean()
    except Exception as e:
        print(f"Error reading Unemployment data: {e}")
        return None

def main():
    gdp_file = input("GDP File Name (e.g. gdp.txt): ")
    uemp_file = input("Unemployment File Name (e.g. uemp.csv): ")

    while True:
        try:
            year = int(input("Year to examine (1948-2008): "))
            if 1948 <= year <= 2008:
                break
            else:
                print("Bad year, try again.")
        except ValueError:
            print("Please enter a valid year.")

    avg_gdp = read_gdp(gdp_file, year)
    avg_uemp = read_unemployment(uemp_file, year)

    if avg_gdp is not None and avg_uemp is not None:
        print(f"For {year}, average GDP: {avg_gdp:.3f} and average unemployment: {avg_uemp:.3f}")

if __name__ == "__main__":
    main()
```

```python
import random

def scramble_word(word):
    punctuation = ''
    if word[-1] in '.,!?;:':
        punctuation = word[-1]
        word = word[:-1]

    if len(word) <= 3:
        return word + punctuation

    middle = list(word[1:-1])
    random.shuffle(middle)
    scrambled = word[0] + ''.join(middle) + word[-1]

    return scrambled + punctuation

def main():
    input_filename = input("Enter the input file name (e.g., input.txt): ")
    output_filename = input("Enter the output file name (e.g., output.txt): ")

    try:
        with open(input_filename, 'r') as infile, open(output_filename, 'w') as outfile:
            for line in infile:
                words = line.strip().split()
                scrambled_line = ' '.join([scramble_word(word) for word in words])
                outfile.write(scrambled_line + '\n')
        print(f"Scrambled text written to '{output_filename}'.")
    except FileNotFoundError:
        print("File not found. Please check the filename and try again.")

main()
```

```python
def read_cipher_text(filename):
    with open(filename, "r") as f:
        words = ""
        for line in f:
            print(line.strip())
            words += line.strip() + " "
    print("\nFull Text:")
    print(words)
    return words

def count_letters(text):
    text = text.lower()
    counts = [0] * 26
    for char in text:
        if char.isalpha():
            index = ord(char) - ord('a')
            counts[index] += 1
    return counts

def get_shift(counts):
    max_index = counts.index(max(counts))
    most_common_char = chr(max_index + ord('a'))
    assumed_shift = (max_index - (ord('e') - ord('a'))) % 26
    return assumed_shift

def decrypt(text, shift):
    decrypted = ""
    for char in text:
        if char.isalpha():
            new_char = chr(((ord(char.lower()) - ord('a') - shift) % 26) + ord('a'))
            decrypted += new_char
        else:
            decrypted += char
    return decrypted

def main():
    filename = input("Enter the cipher text file name (e.g., cipher1.txt): ")
    text = read_cipher_text(filename)
    counts = count_letters(text)
    shift = get_shift(counts)
    print(f"\nDetected Caesar shift: {shift}")
    decoded = decrypt(text, shift)
    print("\nDecrypted Text:\n")
    print(decoded)

main()
```

```python
from io import StringIO

data = """10
0 1
0 2
0 3
1 4
1 6
1 7
1 9
2 3
2 6
2 8
2 9
3 8
3 9
4 6
4 7
4 8
5 9
6 8
7 8
8 9
"""

fp = StringIO(data)
def read_file(fp):
    n = int(fp.readline().strip())
    network = [[] for _ in range(n)]
    for line in fp:
        u, v = map(int, line.strip().split())
        network[u].append(v)
        network[v].append(u)
    return network

def init_matrix(n):
    return [[0]*n for _ in range(n)]

def num_in_common_between_lists(list1, list2):
    return len(set(list1) & set(list2))

def calc_similarity_scores(network):
    n = len(network)
    sim = init_matrix(n)
    for i in range(n):
        for j in range(n):
            if i != j:
                sim[i][j] = num_in_common_between_lists(network[i], network[j])
    return sim

def recommend(user_id, network, similarity_matrix):
    similarities = similarity_matrix[user_id]
    max_common = -1
    friend_id = -1
    for idx, score in enumerate(similarities):
        if idx != user_id and idx not in network[user_id]:
            if score > max_common:
                max_common = score
                friend_id = idx
    return friend_id

network = read_file(fp)
similarity_matrix = calc_similarity_scores(network)

print("Network:")
for i, friends in enumerate(network):
    print(f"{i} : {friends}")

print("\nSimilarity Matrix:")
for row in similarity_matrix:
    print(row)

for test_user in [0, 3, 8]:
    suggestion = recommend(test_user, network, similarity_matrix)
    print(f"\nSuggested friend for user {test_user} is: {suggestion}")
```

```python
import csv

def get_input_descriptor():
    while True:
        try:
            filename = input("Open what file: ")
            file = open(filename, 'r')
            return file
        except FileNotFoundError:
            print("Bad file name, try again")

def get_data_list(file_object, column_number):
    reader = csv.reader(file_object)
    next(reader)
    result = []

    for row in reader:
        try:
            date = row[0]
            value = float(row[column_number])
            result.append((date, value))
        except (IndexError, ValueError):
            continue

    return result

def average_data(list_of_tuples):
    monthly_data = {}

    for date, value in list_of_tuples:
        year_month = date[:7]
        if year_month in monthly_data:
            monthly_data[year_month].append(value)
        else:
            monthly_data[year_month] = [value]

    average_list = []
    for month, values in monthly_data.items():
        avg = sum(values) / len(values)
        formatted_date = f"{month[5:7]}-{month[0:4]}"
        average_list.append((avg, formatted_date))

    return average_list

def main():
    file = get_input_descriptor()
    column = int(input("What column: "))
    data_list = get_data_list(file, column)
    file.close()
    monthly_avg = average_data(data_list)
    sorted_avg = sorted(monthly_avg)

    print(f"Lowest 6 for column {column}")
    for val, date in sorted_avg[:6]:
        print(f"Date:{date}, Value: {val:.2f}")

    print(f"\nHighest 6 for column {column}")
    for val, date in sorted_avg[-6:][::-1]:
        print(f"Date:{date}, Value: {val:.2f}")

main()
```

```python
import random
import re

def scramble_word(word_str):
    if len(word_str) <= 3:
        return word_str
    match = re.match(r"(\w)(\w+)(\w)(\W*)", word_str)
    if not match:
        return word_str

    first, middle, last, punct = match.groups()

    if len(middle) <= 1:
        return word_str

    middle_list = list(middle)
    random.shuffle(middle_list)
    scrambled = first + ''.join(middle_list) + last + punct
    return scrambled

def scramble_line(line_str):
    words = line_str.split()
    scrambled_words = [scramble_word(word) for word in words]
    return ' '.join(scrambled_words)

def open_read_file(file_name_str):
    while True:
        try:
            file = open(file_name_str, 'r')
            return file
        except FileNotFoundError:
            file_name_str = input("Bad file name, try again: ")

def main():
    out_file_name = input("Output file name: ")
    in_file_name = input("Input file name: ")

    in_file = open_read_file(in_file_name)
    out_file = open(out_file_name, 'w')

    for line in in_file:
        scrambled = scramble_line(line)
        out_file.write(scrambled + '\n')

    in_file.close()
    out_file.close()
    print(f"Scrambled content written to {out_file_name}")

main()
```

```python
def calculate_efficiency(data):
    try:
        gp = int(data[6])
        pts = int(data[17])
        reb = int(data[11])
        ast = int(data[12])
        stl = int(data[13])
        blk = int(data[14])
        fga = int(data[7])
        fgm = int(data[8])
        fta = int(data[9])
        ftm = int(data[10])
        to = int(data[15])

        if gp == 0:
            return 0
        eff = (pts + reb + ast + stl + blk - ((fga - fgm) + (fta - ftm) + to)) / gp
        return round(eff, 2)
    except:
        return 0

def main():
    input_filename = 'player_regular_season_career.txt'
    output_filename = 'top50.txt'

    with open(input_filename, 'r') as file:
        lines = file.readlines()

    players = []
    most_mins = most_games = most_pts = most_reb = most_fouls = most_ft = ("", "", 0)

    for line in lines[1:]:
        data = line.strip().split('|')
        if len(data) < 18:
            continue

        first, last = data[1], data[2]
        eff = calculate_efficiency(data)
        players.append([eff, first, last])

        try:
            mins = int(data[5])
            games = int(data[6])
            pts = int(data[17])
            reb = int(data[11])
            pf = int(data[16])
            ftm = int(data[10])
        except ValueError:
            continue

        if mins > most_mins[2]: most_mins = (first, last, mins)
        if games > most_games[2]: most_games = (first, last, games)
        if pts > most_pts[2]: most_pts = (first, last, pts)
        if reb > most_reb[2]: most_reb = (first, last, reb)
        if pf > most_fouls[2]: most_fouls = (first, last, pf)
        if ftm > most_ft[2]: most_ft = (first, last, ftm)

    players.sort(reverse=True)

    with open(output_filename, 'w') as out_file:
        for p in players[:50]:
            out_file.write(f"{p[1]} {p[2]} {p[0]}\n")

    print(f"Most minutes played: {most_mins[1]}, {most_mins[0]}, {most_mins[2]}")
    print(f"Most games played: {most_games[1]}, {most_games[0]}, {most_games[2]}")
    print(f"Most points scored: {most_pts[1]}, {most_pts[0]}, {most_pts[2]}")
    print(f"Most rebounds: {most_reb[1]}, {most_reb[0]}, {most_reb[2]}")
    print(f"Most penalties: {most_fouls[1]}, {most_fouls[0]}, {most_fouls[2]}")
    print(f"Most freethrows made: {most_ft[1]}, {most_ft[0]}, {most_ft[2]}")

main()
```

```python
def read_word_parts(filename):
    with open(filename, 'r') as file:
        category = None
        parts = {'prefix': [], 'root': [], 'suffix': []}
        for line in file:
            line = line.strip()
            if not line or line.startswith("#"):
                if line.lower().startswith('#prefix'):
                    category = 'prefix'
                elif line.lower().startswith('#root'):
                    category = 'root'
                elif line.lower().startswith('#suffix'):
                    category = 'suffix'
                continue
            if category:
                parts[category].append([line, 0])
    return parts

def read_word_list(filename):
    with open(filename, 'r') as file:
        return [line.strip().lower() for line in file if line.strip()]

def tally_parts(parts, word_list):
    for word in word_list:
        for entry in parts['prefix']:
            if word.startswith(entry[0]):
                entry[1] += 1
        for entry in parts['root']:
            if entry[0] in word:
                entry[1] += 1
        for entry in parts['suffix']:
            if word.endswith(entry[0]):
                entry[1] += 1

def print_results(parts):
    print("Roots:")
    for root, count in parts['root']:
        print("%-15s %5d" % (root, count))
    print("\nPrefixes:")
    for prefix, count in parts['prefix']:
        print("%-15s %5d" % (prefix, count))
    print("\nSuffixes:")
    for suffix, count in parts['suffix']:
        print("%-15s %5d" % (suffix, count))

def main():
    word_parts_file = "rootsPrefixesSuffixes.txt"
    word_list_file = "longWordList.txt"

    parts = read_word_parts(word_parts_file)
    word_list = read_word_list(word_list_file)

    tally_parts(parts, word_list)
    print_results(parts)

main()
```

```python
def read_words_from_file():
    while True:
        filename = input("Enter word list filename: ")
        try:
            with open(filename, 'r') as f:
                return [line.strip().lower() for line in f if line.strip()]
        except FileNotFoundError:
            print("File not found, try again.")

def find_anagram_groups(words):
    anagram_dict = {}
    for word in words:
        key = ''.join(sorted(word))
        anagram_dict.setdefault(key, []).append(word)
    return anagram_dict

def report_results(anagram_dict):
    count_by_size = {}
    max_group_size = 0
    max_groups = []

    for group in anagram_dict.values():
        size = len(group)
        if size < 2:
            continue
        count_by_size[size] = count_by_size.get(size, 0) + 1
        if size > max_group_size:
            max_group_size = size
            max_groups = [group]
        elif size == max_group_size:
            max_groups.append(group)

    print("Anagram group sizes (excluding largest group):")
    for size in sorted(count_by_size):
        if size < max_group_size:
            print(f"{size}-word groups: {count_by_size[size]}")

    print(f"\nLargest anagram group(s) with {max_group_size} words:")
    for group in max_groups:
        print(" -> ".join(sorted(group)))

def main():
    words = read_words_from_file()
    anagram_dict = find_anagram_groups(words)
    report_results(anagram_dict)

main()
```

```python
def get_data_list(file_name):
    data_list = []
    with open(file_name, 'r') as f:
        next(f)
        for line in f:
            parts = line.strip().split(',')
            if len(parts) < 7:
                continue
            data_list.append(parts)
    return data_list

def get_monthly_averages(data_list):
    monthly_data = {}

    for entry in data_list:
        date, _, _, _, _, volume, adj_close = entry[0], entry[1], entry[2], entry[3], entry[4], entry[5], entry[6]
        try:
            vol = float(volume)
            close = float(adj_close)
            key = date[:7]
            if key not in monthly_data:
                monthly_data[key] = [0.0, 0.0]
            monthly_data[key][0] += vol * close
            monthly_data[key][1] += vol
        except ValueError:
            continue

    monthly_averages = []
    for key in monthly_data:
        total_sales, total_volume = monthly_data[key]
        if total_volume > 0:
            avg = total_sales / total_volume
            formatted_key = f"{key[5:7]}-{key[0:4]}"
            monthly_averages.append((round(avg, 2), formatted_key))

    return monthly_averages

def print_info(monthly_averages_list):
    with open("monthly_averages.txt", "w") as f:
        sorted_list = sorted(monthly_averages_list, reverse=True)

        f.write("6 best months for google stock:\n")
        for avg, date in sorted_list[:6]:
            f.write(f"{date}, {avg:.2f}\n")

        f.write("\n6 worst months for Google stock:\n")
        for avg, date in sorted_list[-6:]:
            f.write(f"{date}, {avg:.2f}\n")

def main():
    file_name = "table.csv"
    data = get_data_list(file_name)
    monthly_averages = get_monthly_averages(data)
    print_info(monthly_averages)

main()
```

```python
def generate_pascals_triangle(height):
    triangle = []

    for row_num in range(height):
        row = [1] * (row_num + 1)
        for j in range(1, row_num):
            row[j] = triangle[row_num - 1][j - 1] + triangle[row_num - 1][j]
        triangle.append(row)
    return triangle

def print_triangle(triangle):
    height = len(triangle)
    for row in triangle:
        row_str = ' '.join(str(num) for num in row)
        print(row_str.center(height * 4))

def main():
    while True:
        try:
            height = int(input("Enter the height of Pascal's Triangle (positive integer): "))
            if height > 0:
                break
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

    triangle = generate_pascals_triangle(height)
    print_triangle(triangle)

if __name__ == "__main__":
    main()
```

```python
import math

PI = math.pi
FLUID_OUNCE_TO_CUBIC_INCH = 1.805
DOZEN = 12

def fluid_volume_to_cubic_inches(volume_oz):
    return volume_oz * FLUID_OUNCE_TO_CUBIC_INCH

def inner_radius(volume_cubic_in, height):
    return math.sqrt(volume_cubic_in / (PI * height))

def outer_radius(inner_r, thickness):
    return inner_r + thickness

def outer_height(fluid_height, thickness):
    return fluid_height + thickness

def glass_volume(Ro, Ri, height):
    return PI * height * (Ro**2 - Ri**2)

def glass_weight(volume, density):
    return volume * density

def box_weight(single_glass_weight):
    return (single_glass_weight * DOZEN) / 16

def box_arrange(diameter, height):
    best = None
    for r in range(1, DOZEN + 1):
        for c in range(1, DOZEN + 1):
            if r * c > DOZEN:
                continue
            s = (DOZEN + r * c - 1) // (r * c)
            L = c * diameter
            W = r * diameter
            H = s * height
            volume = L * W * H
            if best is None or volume < best[1]:
                best = ((r, c, s), volume, (L, W, H))
    return best[0], best[2]

def process_glass_line(volume, height, density, thickness):
    vol_cubic_in = fluid_volume_to_cubic_inches(volume)
    Ri = inner_radius(vol_cubic_in, height)
    Ro = outer_radius(Ri, thickness)
    H = outer_height(height, thickness)
    D = 2 * Ro
    V = glass_volume(Ro, Ri, H)
    W = glass_weight(V, density)
    layout, dimensions = box_arrange(D, H)
    total_weight_lb = box_weight(W)
    return volume, height, density, thickness, layout, dimensions, round(total_weight_lb, 2)

def main():
    with open("test.txt", "r") as file:
        for line in file:
            volume, height, density, thickness = map(float, line.strip().split())
            v, h, d, t, layout, dim, wt = process_glass_line(volume, height, density, thickness)
            print(f"Input: Volume = {v} oz, Height = {h} in, Density = {d} oz/in¬≥, Thickness = {t} in")
            print(f"Arrangement: {layout[0]} x {layout[1]} x {layout[2]}")
            print(f"Box Dimensions: {dim[0]:.2f} x {dim[1]:.2f} x {dim[2]:.2f} inches")
            print(f"Total Weight: {wt} lbs\n")

if __name__ == "__main__":
    main()
```

```python
import string

def open_file():
    while True:
        try:
            filename = input("Enter a file name: ")
            fp = open(filename, "r")
            return fp
        except FileNotFoundError:
            print("Error -- file not found. Try again.")

def read_data(fp):
    data_dict = {}
    for idx, line in enumerate(fp, 1):
        line = line.strip().lower()
        for p in string.punctuation:
            if p in ["'", "-"]:
                line = line.replace(p, "")
            else:
                line = line.replace(p, " ")
        words = line.split()
        for word in words:
            if word.isalpha() and len(word) >= 2:
                if word not in data_dict:
                    data_dict[word] = {idx}
                else:
                    data_dict[word].add(idx)
    return data_dict

def find_cooccurance(D, inp_str):
    words = inp_str.lower().split()
    found_lines = None
    for word in words:
        if word in D:
            if found_lines is None:
                found_lines = D[word].copy()
            else:
                found_lines &= D[word]
        else:
            return []
    return sorted(found_lines) if found_lines else []

def main():
    fp = open_file()
    data_dict = read_data(fp)
    while True:
        user_input = input("Enter space-separated words (or 'q' to quit): ")
        if user_input.lower() == "q":
            break
        result = find_cooccurance(data_dict, user_input)
        print("The co-occurance for:", user_input)
        if result:
            print("Lines:", ", ".join(map(str, result)))
        else:
            print("Lines: None.")

if __name__ == "__main__":
    main()
```

```python
import string

def read_documents(filename):
    with open(filename, 'r') as file:
        documents = []
        current_doc = ''
        for line in file:
            if line.strip() == "<NEW DOCUMENT>":
                if current_doc:
                    documents.append(current_doc.strip())
                    current_doc = ''
            else:
                current_doc += line
        if current_doc:
            documents.append(current_doc.strip())
    return documents

def build_index(documents):
    word_dict = {}
    for i, doc in enumerate(documents, 1):
        words = doc.lower().split()
        for word in words:
            word = word.strip(string.punctuation)
            if word.isalpha():
                if word not in word_dict:
                    word_dict[word] = {i}
                else:
                    word_dict[word].add(i)
    return word_dict

def search_documents(index, query):
    terms = query.lower().split()
    result = None
    for term in terms:
        term = term.strip(string.punctuation)
        if term in index:
            if result is None:
                result = index[term].copy()
            else:
                result &= index[term]
        else:
            return set()
    return result if result else set()

def main():
    filename = input("Enter filename: ")
    docs = read_documents(filename)
    index = build_index(docs)

    while True:
        print("\nMenu:")
        print("1. Search")
        print("2. Display document")
        print("3. Quit")
        choice = input("Enter choice: ")

        if choice == '1':
            query = input("Enter search words: ")
            results = search_documents(index, query)
            if results:
                print("Documents found:", sorted(results))
            else:
                print("No relevant documents found.")

        elif choice == '2':
            try:
                doc_num = int(input(f"Enter document number (1-{len(docs)}): "))
                if 1 <= doc_num <= len(docs):
                    print(f"\nDocument {doc_num}:\n{docs[doc_num - 1]}")
                else:
                    print("Invalid document number.")
            except ValueError:
                print("Invalid input. Please enter a number.")

        elif choice == '3':
            print("Exiting.")
            break

        else:
            print("Invalid option. Try again.")

if __name__ == "__main__":
    main()
```

```python
def process_input_file(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
    except FileNotFoundError:
        print("File not found.")
        return {}, {}

    editor_dict = {}
    article_dict = {}

    for line in lines:
        line = line.strip()
        if line.startswith("REVISION"):
            parts = line.split()
            if len(parts) < 6:
                continue
            article = parts[3]
            editor = parts[5]

            if editor not in editor_dict:
                editor_dict[editor] = set()
            editor_dict[editor].add(article)

            if article not in article_dict:
                article_dict[article] = [0, set()]
            article_dict[article][0] += 1
            article_dict[article][1].add(editor)

    return editor_dict, article_dict

def top_n_editors(editor_dict, n):
    print(f"Top {n} editors:")
    print(f"{'Editor':<30}{'Articles'}")
    sorted_items = sorted(editor_dict.items(), key=lambda x: (-len(x[1]), x[0]))
    for editor, articles in sorted_items[:n]:
        print(f"{editor:<30}{len(articles)}")

def top_n_edits(article_dict, n):
    print(f"Top {n} articles by number of edits:")
    print(f"{'Article':<30}{'Edits'}")
    sorted_items = sorted(article_dict.items(), key=lambda x: (-x[1][0], x[0]))
    for article, (count, _) in sorted_items[:n]:
        print(f"{article:<30}{count}")

def top_n_articles(article_dict, n):
    print(f"Top {n} articles by number of editors:")
    print(f"{'Article':<30}{'Editors'}")
    sorted_items = sorted(article_dict.items(), key=lambda x: (-len(x[1][1]), x[0]))
    for article, (_, editors) in sorted_items[:n]:
        print(f"{article:<30}{len(editors)}")

def show_help():
    print("Available commands:")
    print("INPUT filename")
    print("TOP n EDITORS")
    print("TOP n EDITS")
    print("TOP n ARTICLES")
    print("HELP")
    print("QUIT")

def main():
    editor_dict = {}
    article_dict = {}

    while True:
        command = input("Enter command: ").strip()
        parts = command.split()

        if not parts:
            continue
        cmd = parts[0].upper()

        if cmd == "QUIT":
            break
        elif cmd == "HELP":
            show_help()
        elif cmd == "INPUT" and len(parts) == 2:
            editor_dict, article_dict = process_input_file(parts[1])
            print("File processed.")
        elif cmd == "TOP" and len(parts) == 3 and parts[1].isdigit():
            n = int(parts[1])
            category = parts[2].upper()
            if category == "EDITORS":
                top_n_editors(editor_dict, n)
            elif category == "EDITS":
                top_n_edits(article_dict, n)
            elif category == "ARTICLES":
                top_n_articles(article_dict, n)
            else:
                print("Unknown TOP category. Try again.")
        else:
            print("Invalid command. Type HELP for a list of commands.")

if __name__ == "__main__":
    main()
```

```python
import string

def clean_word(word):
    """Remove punctuation and convert to lowercase."""
    return word.strip(string.punctuation).lower()

def fill_completions(c_dict, fd):
    """Fill the dictionary with word positions."""
    for line in fd:
        for word in line.split():
            word = clean_word(word)
            if not word.isalpha() or len(word) <= 1:
                continue
            for i, letter in enumerate(word):
                key = (i, letter)
                if key not in c_dict:
                    c_dict[key] = set()
                c_dict[key].add(word)

def find_completions(prefix, c_dict):
    """Return possible completions using set intersection."""
    prefix = prefix.lower()
    if not prefix.isalpha():
        return set()

    result = None
    for i, letter in enumerate(prefix):
        key = (i, letter)
        if key not in c_dict:
            return set()
        if result is None:
            result = c_dict[key].copy()
        else:
            result &= c_dict[key]
    return result if result else set()

def main():
    filename = "ap_docs.txt"
    try:
        with open(filename, "r") as f:
            c_dict = {}
            fill_completions(c_dict, f)
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
        return

    while True:
        prefix = input("Enter prefix (# to quit): ").strip()
        if prefix == "#":
            break
        completions = find_completions(prefix, c_dict)
        if completions:
            print("Completions:", completions)
        else:
            print("No completions found.")

if __name__ == "__main__":
    main()
```

```python
def build_movie_dict(filename):
    movie_dict = {}
    with open(filename, 'r') as f:
        for line in f:
            parts = [part.strip() for part in line.strip().split(',')]
            actor = parts[0]
            movies = parts[1:]
            for movie in movies:
                if movie not in movie_dict:
                    movie_dict[movie] = set()
                movie_dict[movie].add(actor)
    return movie_dict

def get_coactors(movie_dict, actor_name):
    coactors = set()
    for actors in movie_dict.values():
        if actor_name in actors:
            coactors.update(actors)
    coactors.discard(actor_name)
    return coactors

def movie_operation(movie_dict, movie1, movie2, operator):
    if movie1 not in movie_dict or movie2 not in movie_dict:
        print("One or both movie titles not found.")
        return set()

    set1 = movie_dict[movie1]
    set2 = movie_dict[movie2]

    if operator == '&':
        return set1 & set2
    elif operator == '|':
        return set1 | set2
    elif operator == '^':
        return set1 ^ set2
    else:
        print("Invalid operator.")
        return set()

def main():
    filename = "movies.txt"
    movie_dict = build_movie_dict(filename)

    while True:
        print("\nMenu:")
        print("1. Perform set operation on two movies")
        print("2. Find co-actors of an actor")
        print("3. Quit")
        choice = input("Enter your choice (1/2/3): ").strip()

        if choice == '1':
            user_input = input("Enter two movie names separated by &, | or ^: ")
            for op in ['&', '|', '^']:
                if op in user_input:
                    movie1, movie2 = [m.strip() for m in user_input.split(op)]
                    result = movie_operation(movie_dict, movie1, movie2, op)
                    print("Resulting actors:")
                    for actor in sorted(result):
                        print(actor)
                    break
            else:
                print("Invalid input. No operator found.")

        elif choice == '2':
            actor_name = input("Enter actor's name: ").strip()
            coactors = get_coactors(movie_dict, actor_name)
            if coactors:
                print(f"Co-actors of {actor_name}:")
                for coactor in sorted(coactors):
                    print(coactor)
            else:
                print(f"No co-actors found for {actor_name}.")

        elif choice == '3':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```

```python
import math

def compute_regression(x, y):
    n = len(x)
    sumX = sum(x)
    sumY = sum(y)
    sumXY = sum([xi * yi for xi, yi in zip(x, y)])
    sumXsq = sum([xi ** 2 for xi in x])
    slope = (n * sumXY - sumX * sumY) / (n * sumXsq - sumX ** 2)
    intercept = (sumY - slope * sumX) / n
    return slope, intercept

def compute_correlation(x, y):
    n = len(x)
    sumX = sum(x)
    sumY = sum(y)
    sumXY = sum([xi * yi for xi, yi in zip(x, y)])
    sumXsq = sum([xi ** 2 for xi in x])
    sumYsq = sum([yi ** 2 for yi in y])
    numerator = n * sumXY - sumX * sumY
    denominator = math.sqrt((n * sumXsq - sumX ** 2) * (n * sumYsq - sumY ** 2))
    return numerator / denominator

def calculate_bmi(weight, height_cm):
    height_m = height_cm / 100
    if height_m == 0:
        return None
    return weight / (height_m ** 2)

def weight_formula(cd, cdepth, bitro, wrist, ankle, height):
    return -110 + 1.34 * cd + 1.54 * cdepth + 1.20 * bitro + 1.11 * wrist + 1.15 * ankle + 0.177 * height

def read_data(filename):
    with open(filename, 'r') as f:
        age_list = []
        bmi_list = []
        weight_list = []
        formula_list = []
        for line in f:
            parts = list(map(float, line.split()))
            age = parts[22]
            weight = parts[23]
            height = parts[24]

            if height == 0:
                continue

            chest_diameter = parts[4]
            chest_depth = parts[3]
            bitrochanteric = parts[10]
            wrist_girth = parts[20]
            ankle_girth = parts[18]

            bmi = calculate_bmi(weight, height)
            if bmi is None:
                continue

            predicted = weight_formula(chest_diameter, chest_depth, bitrochanteric, wrist_girth, ankle_girth, height)

            age_list.append(age)
            bmi_list.append(bmi)
            weight_list.append(weight)
            formula_list.append(predicted)
        return age_list, bmi_list, weight_list, formula_list

def main():
    age, bmi, weight, predicted = read_data("body.dat.txt")

    print("BMI vs Age")
    slope1, intercept1 = compute_regression(age, bmi)
    print(f"Slope: {slope1:.4f}")
    print(f"Intercept: {intercept1:.4f}")
    corr1 = compute_correlation(age, bmi)
    print(f"Correlation: {corr1:.4f}")

    print("\nPredicted Weight Formula vs Actual Weight")
    slope2, intercept2 = compute_regression(predicted, weight)
    print(f"Slope: {slope2:.4f}")
    print(f"Intercept: {intercept2:.4f}")
    corr2 = compute_correlation(predicted, weight)
    print(f"Correlation: {corr2:.4f}")

if __name__ == "__main__":
    main()
```

```python
ATTRIBUTES = ['radius', 'texture', 'perimeter', 'area', 'smoothness',
              'compactness', 'concavity', 'concave', 'symmetry', 'fractal']

def read_data(filename):
    data = []
    with open(filename, 'r') as file:
        lines = file.readlines()[1:]
        for line in lines:
            parts = line.strip().split(',')
            tumor = {}
            tumor['id'] = parts[0]
            tumor['class'] = parts[1]
            for i, attr in enumerate(ATTRIBUTES):
                tumor[attr] = float(parts[2 + i])
            data.append(tumor)
    return data

def compute_averages(data):
    mal_totals = {attr: 0.0 for attr in ATTRIBUTES}
    ben_totals = {attr: 0.0 for attr in ATTRIBUTES}
    mal_count = 0
    ben_count = 0

    for tumor in data:
        if tumor['class'] == 'M':
            mal_count += 1
            for attr in ATTRIBUTES:
                mal_totals[attr] += tumor[attr]
        else:
            ben_count += 1
            for attr in ATTRIBUTES:
                ben_totals[attr] += tumor[attr]

    mal_avg = {attr: mal_totals[attr] / mal_count for attr in ATTRIBUTES}
    ben_avg = {attr: ben_totals[attr] / ben_count for attr in ATTRIBUTES}

    midpoints = {attr: (mal_avg[attr] + ben_avg[attr]) / 2 for attr in ATTRIBUTES}

    return mal_avg, ben_avg, midpoints

def classify(tumor, midpoints):
    votes = 0
    for attr in ATTRIBUTES:
        if tumor[attr] >= midpoints[attr]:
            votes += 1
    return 'M' if votes >= 5 else 'B'

def calculate_accuracy(test_data, midpoints):
    correct = 0
    for tumor in test_data:
        prediction = classify(tumor, midpoints)
        if prediction == tumor['class']:
            correct += 1
    accuracy = correct / len(test_data) * 100
    print(f"Correct predictions: {correct}/{len(test_data)}")
    print(f"Accuracy: {accuracy:.2f}%")

def check_patients(test_data, midpoints):
    id_map = {tumor['id']: tumor for tumor in test_data}

    while True:
        pid = input("Enter patient ID to check (or 'quit' to stop): ")
        if pid.lower() == 'quit':
            break
        if pid not in id_map:
            print("Patient not found.")
            continue

        tumor = id_map[pid]
        print(f"\nPatient ID: {pid}")
        votes = 0
        for attr in ATTRIBUTES:
            val = tumor[attr]
            cutoff = midpoints[attr]
            result = 'Malignant' if val >= cutoff else 'Benign'
            if val >= cutoff:
                votes += 1
            print(f"{attr:>12}: {val:.3f} vs {cutoff:.3f} ‚Üí {result}")
        final = 'Malignant' if votes >= 5 else 'Benign'
        print(f"Overall prediction: {final}\n")

def main():
    print("Reading training data...")
    training_data = read_data("cancerTrainingData.txt")
    print("Training classifier...")
    mal_avg, ben_avg, midpoints = compute_averages(training_data)

    print("Reading test data...")
    test_data = read_data("cancerTestingData.txt")
    print("Testing classifier...")
    calculate_accuracy(test_data, midpoints)

    print("\n--- Patient Lookup ---")
    check_patients(test_data, midpoints)

if __name__ == "__main__":
    main()
```

```python
import csv

CONTINUOUS_ATTRS = ["age", "education-num", "capital-gain", "capital-loss", "hours-per-week"]
CATEGORICAL_ATTRS = ["workclass", "marital-status", "occupation", "relationship", "race", "sex"]

def read_data(filename):
    """Reads CSV and returns list of dictionaries"""
    with open(filename, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        return list(reader)

def init_model():
    """Initialize empty model structure"""
    model = {}
    for attr in CONTINUOUS_ATTRS:
        model[attr] = 0.0
    for attr in CATEGORICAL_ATTRS:
        model[attr] = {}
    return model

def train_classifier(data):
    """Train the model on labeled data"""
    model = {">50K": init_model(), "<=50K": init_model()}
    counts = {">50K": 0, "<=50K": 0}

    for record in data:
        label = record["class"]
        counts[label] += 1
        for attr in CONTINUOUS_ATTRS:
            model[label][attr] += float(record[attr])
        for attr in CATEGORICAL_ATTRS:
            val = record[attr]
            model[label][attr][val] = model[label][attr].get(val, 0) + 1

    for label in model:
        for attr in CONTINUOUS_ATTRS:
            model[label][attr] /= counts[label]
        for attr in CATEGORICAL_ATTRS:
            total = sum(model[label][attr].values())
            for val in model[label][attr]:
                model[label][attr][val] /= total

    return model

def classify(record, model):
    """Classify a single record using vote system"""
    votes = {">50K": 0, "<=50K": 0}

    for attr in CONTINUOUS_ATTRS:
        val = float(record[attr])
        diff_gt = abs(val - model[">50K"][attr])
        diff_le = abs(val - model["<=50K"][attr])
        if diff_gt < diff_le:
            votes[">50K"] += 1
        else:
            votes["<=50K"] += 1

    for attr in CATEGORICAL_ATTRS:
        val = record[attr]
        gt_ratio = model[">50K"][attr].get(val, 0)
        le_ratio = model["<=50K"][attr].get(val, 0)
        if gt_ratio > le_ratio:
            votes[">50K"] += 1
        else:
            votes["<=50K"] += 1

    return ">50K" if votes[">50K"] > votes["<=50K"] else "<=50K"

def report_accuracy(test_data, model):
    """Compare predictions to actual and report accuracy"""
    correct = 0
    for record in test_data:
        predicted = classify(record, model)
        if predicted == record["class"]:
            correct += 1
    total = len(test_data)
    accuracy = correct / total * 100
    print(f"Correct: {correct} / {total}")
    print(f"Accuracy: {accuracy:.2f}%")

def main():
    training_data = read_data("income-training.csv")
    test_data = read_data("income-test.csv")
    model = train_classifier(training_data)
    report_accuracy(test_data, model)

if __name__ == "__main__":
    main()
```

```python
import os
import glob
import random

def load_vocabulary(filename):
    vocab = {}
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.strip().split(':')
            if len(parts) == 2:
                english, spanish = parts
                vocab[english.strip()] = [word.strip() for word in spanish.split(',')]
    return vocab

def save_missed_words(filename, missed):
    with open(filename, 'w', encoding='utf-8') as file:
        for english, spanish in missed.items():
            file.write(f"{english}: {', '.join(spanish)}\n")

def quiz_user(vocab, num_questions):
    missed = {}
    words = random.sample(list(vocab.keys()), num_questions)
    correct = 0

    for word in words:
        correct_answers = vocab[word]
        print(f"\nEnglish word: {word}")
        print(f"Enter {len(correct_answers)} equivalent Spanish word(s).")
        user_answers = []
        for i in range(len(correct_answers)):
            user_input = input(f"Word [{i+1}]: ").strip().lower()
            user_answers.append(user_input)

        if sorted(user_answers) == sorted([w.lower() for w in correct_answers]):
            print("Correct!")
            correct += 1
        else:
            print("Incorrect.")
            missed[word] = correct_answers

    print(f"\n{correct} out of {num_questions} correct.")
    return missed

def main():
    print("Welcome to the vocabulary quiz program.")
    txt_files = glob.glob("*.txt")

    if not txt_files:
        print("No vocabulary files (.txt) found in the current directory.")
        return

    print("Select one of the following word lists:")
    for f in txt_files:
        print(f"  {f}")

    selected_file = input("Please make your selection: ").strip()
    if selected_file not in txt_files:
        print("Invalid file selection.")
        return

    vocab = load_vocabulary(selected_file)
    print(f"{len(vocab)} entries found.")

    try:
        num = int(input("How many words would you like to be quizzed on? "))
        if num < 1 or num > len(vocab):
            raise ValueError
    except ValueError:
        print("Invalid number of words.")
        return

    missed_words = quiz_user(vocab, num)

    if missed_words:
        output_file = input("Enter an output file (or press enter to quit): ").strip()
        if output_file:
            save_missed_words(output_file, missed_words)
            print(f"Missed words saved to {output_file}.")

if __name__ == "__main__":
    main()
```

```python
def load_data(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            if line.strip():
                fields = line.strip().split(', ')
                data.append(fields)
    return data

def build_model(records):
    model = {
        'age': 0,
        'education-num': 0,
        'capital-gain': 0,
        'capital-loss': 0,
        'hours-per-week': 0,
        'workclass': {},
        'marital-status': {},
        'occupation': {},
        'relationship': {},
        'race': {},
        'sex': {}
    }
    count = len(records)

    for row in records:
        model['age'] += int(row[0])
        model['education-num'] += int(row[4])
        model['capital-gain'] += int(row[10])
        model['capital-loss'] += int(row[11])
        model['hours-per-week'] += int(row[12])

        for key, index in [('workclass', 1), ('marital-status', 5), ('occupation', 6),
                           ('relationship', 7), ('race', 8), ('sex', 9)]:
            value = row[index]
            model[key][value] = model[key].get(value, 0) + 1

    for key in ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']:
        model[key] /= count

    for key in ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'sex']:
        total = sum(model[key].values())
        for val in model[key]:
            model[key][val] /= total

    return model

def train_classifier(data):
    gt50 = [row for row in data if row[-1] == '>50K']
    le50 = [row for row in data if row[-1] == '<=50K']
    return build_model(gt50), build_model(le50)

def classify_record(record, model_gt, model_le):
    score_gt = 0
    score_le = 0

    for key, idx in [('age', 0), ('education-num', 4), ('capital-gain', 10),
                     ('capital-loss', 11), ('hours-per-week', 12)]:
        val = int(record[idx])
        if abs(val - model_gt[key]) < abs(val - model_le[key]):
            score_gt += 1
        else:
            score_le += 1

    for key, idx in [('workclass', 1), ('marital-status', 5), ('occupation', 6),
                     ('relationship', 7), ('race', 8), ('sex', 9)]:
        val = record[idx]
        if model_gt[key].get(val, 0) > model_le[key].get(val, 0):
            score_gt += 1
        else:
            score_le += 1

    return '>50K' if score_gt > score_le else '<=50K'

def test_classifier(test_data, model_gt, model_le):
    correct = 0
    for row in test_data:
        predicted = classify_record(row, model_gt, model_le)
        if predicted == row[-1]:
            correct += 1
    total = len(test_data)
    print(f"Correct: {correct} / {total}")
    print(f"Accuracy: {correct / total * 100:.2f}%")

if __name__ == "__main__":

    from google.colab import files
    uploaded = files.upload()

    train_data = load_data("annual-income-training.data")
    test_data = load_data("annual-income-test.data")

    model_gt, model_le = train_classifier(train_data)
    test_classifier(test_data, model_gt, model_le)
```

```python
import string

def load_stop_words(filename):
    with open(filename, 'r') as file:
        return set(word.strip().lower() for word in file if word.strip())

def build_concordance(filename, stop_words):
    concordance = {}
    with open(filename, 'r') as file:
        lines = file.readlines()

    for line_num, line in enumerate(lines, 1):
        original_line = line.strip()
        clean_line = line.lower().translate(str.maketrans('', '', string.punctuation))
        words = clean_line.strip().split()

        for word in words:
            if word in stop_words:
                continue
            if word not in concordance:
                concordance[word] = {'count': 0, 'occurrences': []}

            concordance[word]['count'] += 1
            upper_context = replace_word_case_insensitive(original_line, word, word.upper())
            concordance[word]['occurrences'].append((line_num, upper_context))

    return concordance

def replace_word_case_insensitive(text, word, replacement):
    import re
    pattern = re.compile(r'\b' + re.escape(word) + r'\b', re.IGNORECASE)
    return pattern.sub(replacement, text, count=1)

def print_concordance(concordance, filename):
    print(f"Concordance for file {filename}")
    for word in sorted(concordance):
        entry = concordance[word]
        print(f"{word} : Total Count: {entry['count']}")
        for line_num, context in entry['occurrences']:
            print(f" Line:{line_num}: {context}")
        print()

from google.colab import files
uploaded = files.upload()
stop_file = input("Enter stop word file (e.g. english.stop.txt): ").strip()
text_file = input("Analyze what file (e.g. itsy_bitsy_spider.txt): ").strip()
stop_words = load_stop_words(stop_file)
concordance = build_concordance(text_file, stop_words)
print_concordance(concordance, text_file)
```

```python
def init_dictionary(file_string, sunspot_dict):
    try:
        with open(file_string, 'r') as file:
            for line in file:
                if line.startswith("#") or not line.strip():
                    continue
                parts = line.strip().split()
                if len(parts) != 3:
                    continue
                year = int(parts[0])
                month = int(parts[1])
                sunspot = float(parts[2])
                if sunspot < 0:
                    continue
                if year not in sunspot_dict:
                    sunspot_dict[year] = [None] * 12
                sunspot_dict[year][month - 1] = sunspot
    except FileNotFoundError:
        raise IOError("File not found.")

def avg_sunspot(sunspot_dict, year_tuple=(1749, 2011), month_tuple=(1, 12)):
    start_year, end_year = year_tuple
    start_month, end_month = month_tuple

    if start_year < 1749 or end_year > 2011 or start_year > end_year:
        raise KeyError("Invalid year range.")
    if start_month < 1 or end_month > 12 or start_month > end_month:
        raise IndexError("Invalid month range.")

    total = 0
    count = 0
    for year in range(start_year, end_year + 1):
        if year in sunspot_dict:
            for month in range(start_month - 1, end_month):
                value = sunspot_dict[year][month]
                if value is not None:
                    total += value
                    count += 1
    return total / count if count > 0 else 0

if __name__ == "__main__":
    sunspot_dict = {}
    data_file = "sunspot_data.txt"

    init_dictionary(data_file, sunspot_dict)
    avg_all = avg_sunspot(sunspot_dict)
    print(f"Average sunspot (1749-2011, Jan-Dec): {avg_all:.2f}")
    avg_1749_q1 = avg_sunspot(sunspot_dict, (1749, 1749), (1, 3))
    print(f"Average sunspot (1749, Jan-Mar): {avg_1749_q1:.2f}")
    avg_2011_q1 = avg_sunspot(sunspot_dict, (2011, 2011), (1, 3))
    print(f"Average sunspot (2011, Jan-Mar): {avg_2011_q1:.2f}")
```

```python
def load_reference_words(filename):
    with open(filename, 'r') as file:
        return set(word.strip().lower() for word in file)

def get_candidates(word, reference_words):
    candidates = []
    for ref_word in reference_words:
        if abs(len(word) - len(ref_word)) <= 2 and word[0] == ref_word[0]:
            candidates.append(ref_word)
    return candidates

def spell_check_file(filename, ref_words):
    output_lines = []
    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()
            new_line = []
            for word in words:
                stripped_word = ''.join(filter(str.isalpha, word)).lower()
                if stripped_word == '' or stripped_word in ref_words:
                    new_line.append(word)
                else:
                    print(f"\n{stripped_word} : ", end='')
                    candidates = get_candidates(stripped_word, ref_words)
                    for i, cand in enumerate(candidates):
                        print(f"{i}:{cand}", end=', ')
                    print()

                    while True:
                        user_input = input("Action (-1 leave as is, -2 manual input, or 0-n, q to quit): ")
                        if user_input.lower() in ['q', 'quit', 'exit']:
                            print("Exiting program.")
                            exit()
                        try:
                            choice = int(user_input)
                            if choice == -1 or choice == -2 or (0 <= choice < len(candidates)):
                                break
                            else:
                                print("Invalid choice. Please enter a valid index, -1, -2, or 'q' to quit.")
                        except ValueError:
                            print("Invalid input. Please enter a number or 'q' to quit.")

                    if choice == -1:
                        new_line.append(word)
                    elif choice == -2:
                        replacement = input("Replacement: ")
                        new_line.append(replacement)
                    else:
                        new_line.append(candidates[choice])
            output_lines.append(' '.join(new_line))
    return output_lines

def main():
    ref_file = input("What reference list? (default: ref.txt): ").strip()
    if ref_file == "":
        ref_file = "ref.txt"
    input_file = input("What file to check?: ").strip()
    try:
        reference_words = load_reference_words(ref_file)
        corrected_lines = spell_check_file(input_file, reference_words)
        output_file = input_file.rsplit('.', 1)
        if len(output_file) == 2:
            output_filename = output_file[0] + "-chk." + output_file[1]
        else:
            output_filename = input_file + "-chk"
        with open(output_filename, 'w') as out_file:
            for line in corrected_lines:
                out_file.write(line + '\n')
        print("All done. Output saved to", output_filename)
    except FileNotFoundError as e:
        print("File error:", e)

if __name__ == "__main__":
    main()
```

```python
def read_web_file(filename):
    unigram_counts = {}
    bigram_counts = {}
    total_bigrams = 0

    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()[1:]
            for word in words:
                unigram_counts[word] = unigram_counts.get(word, 0) + 1

            for i in range(len(words) - 1):
                pair = tuple(sorted([words[i], words[i + 1]]))
                bigram_counts[pair] = bigram_counts.get(pair, 0) + 1
                total_bigrams += 1

    return unigram_counts, bigram_counts, total_bigrams

def read_bigram_list(filename):
    bigram_list = []
    with open(filename, 'r') as file:
        for line in file:
            word1, word2 = line.strip().split()
            bigram_list.append(tuple(sorted([word1, word2])))
    return bigram_list

def print_contingency_tables(bigram_list, unigrams, bigrams, total_bigram_count):
    for pair in bigram_list:
        w1, w2 = pair
        bigram_count = bigrams.get(pair, 0)
        w1_count = unigrams.get(w1, 0)
        w2_count = unigrams.get(w2, 0)

        w1_only = w1_count - bigram_count
        w2_only = w2_count - bigram_count
        neither = total_bigram_count - (bigram_count + w1_only + w2_only)

        print(f"\n{'':25}{w2:15} ~{w2}")
        print(f"{w1:25}{bigram_count:<15}{w1_only}")
        print(f"~{w1:24}{w2_only:<15}{neither}")
        print("-" * 63)

def main():
    web_file = 'web.txt'
    bigram_file = 'bigrams.txt'

    unigrams, bigrams, total_bigram_count = read_web_file(web_file)
    bigram_list = read_bigram_list(bigram_file)
    print_contingency_tables(bigram_list, unigrams, bigrams, total_bigram_count)

if __name__ == '__main__':
    main()
```

```python
import random

def read_file_words(filename):
    with open(filename, 'r') as file:
        text = file.read()
    return text.split()

def build_markov_chain(words):
    chain = {}
    for i in range(len(words) - 2):
        key = words[i] + ' ' + words[i + 1]
        next_word = words[i + 2]
        if key in chain:
            chain[key].append(next_word)
        else:
            chain[key] = [next_word]
    return chain

def generate_text(words, chain, max_words=500):
    if len(words) < 3:
        return ""
    key = words[0] + ' ' + words[1]
    result = [words[0], words[1]]
    for _ in range(max_words - 2):
        if key not in chain:
            break
        next_word = random.choice(chain[key])
        result.append(next_word)
        key = key.split()[1] + ' ' + next_word
    return ' '.join(result)

def wrap_text(text, width=80):
    words = text.split()
    lines = []
    line = ""
    for word in words:
        if len(line) + len(word) + 1 > width:
            lines.append(line)
            line = word
        else:
            if line:
                line += ' ' + word
            else:
                line = word
    if line:
        lines.append(line)
    return '\n'.join(lines)

def main():
    filename = input("Enter the text file name: ").strip()
    try:
        words = read_file_words(filename)
        chain = build_markov_chain(words)
        text = generate_text(words, chain)
        print("\nThe results:\n" + "-" * 70)
        print(wrap_text(text))
        print("-" * 70)
        output = input("\nEnter file name to write output to <Enter for skip>: ").strip()
        if output:
            with open(output, 'w') as f:
                f.write(wrap_text(text))
            print("Text saved to", output)
    except FileNotFoundError:
        print("File not found.")

if __name__ == "__main__":
    main()
```

```python
def read_presidents(f):
    p = []
    for l in open(f):
        s = l.strip().split(',')
        if len(s) == 3:
            name, yrs, party = s
        else:
            name = s[0].strip() + ', ' + s[1].strip()
            yrs, party = s[2], s[3]
        y1, y2 = map(int, yrs.split('-'))
        p.append({'name': name.strip(), 'start': y1, 'end': y2 - 1, 'party': party.strip()})
    return p

def read_employment(f):
    d = {}
    lines = open(f).readlines()[1:]
    for line in lines:
        parts = line.strip().split(',')
        y = int(parts[0])
        for i in range(1, 13):
            d[(y, i)] = int(parts[i])
    return d

def first_last(emp, pres):
    f = emp.get((pres['start'], 1), 0)
    l = emp.get((pres['end'], 12), 0)
    return f, l

def show_data(title, emp, pres):
    print(f"\n{title} Employment by president")
    print(f"{'Name':30s} {'First':>10} {'Last':>10} {'Change':>10} {'%Change':>9}")
    for p in pres:
        f, l = first_last(emp, p)
        c = l - f
        pct = (c / f * 100) if f else 0
        print(f"{p['name']:30s} {f:10,} {l:10,} {c:10,} {pct:9.1f}%")

pres_file = "/content/presidents.txt"
priv_file = "/content/private_employment.txt"
gov_file = "/content/government_employment.txt"

pres = read_presidents(pres_file)
priv = read_employment(priv_file)
gov = read_employment(gov_file)

show_data("Private", priv, pres)
show_data("Government", gov, pres)
```

```python
import cards

def get_rank_dict(cards_list):
    rank_dict = {}
    for card in cards_list:
        rank = card.get_rank()
        rank_dict[rank] = rank_dict.get(rank, 0) + 1
    return rank_dict

def get_suit_dict(cards_list):
    suit_dict = {}
    for card in cards_list:
        suit = card.get_suit()
        if suit not in suit_dict:
            suit_dict[suit] = []
        suit_dict[suit].append(card)
    return suit_dict

def four_of_a_kind(cards_list):
    rank_dict = get_rank_dict(cards_list)
    for rank, count in rank_dict.items():
        if count == 4:
            return [card for card in cards_list if card.get_rank() == rank]
    return []

def full_house(cards_list):
    rank_dict = get_rank_dict(cards_list)
    three = None
    two = None
    for rank, count in rank_dict.items():
        if count >= 3 and not three:
            three = rank
        elif count >= 2:
            two = rank
    if three and two:
        return [card for card in cards_list if card.get_rank() in (three, two)][:5]
    return []

def three_of_a_kind(cards_list):
    rank_dict = get_rank_dict(cards_list)
    for rank, count in rank_dict.items():
        if count == 3:
            return [card for card in cards_list if card.get_rank() == rank]
    return []

def two_pair(cards_list):
    rank_dict = get_rank_dict(cards_list)
    pairs = [rank for rank, count in rank_dict.items() if count >= 2]
    if len(pairs) >= 2:
        result = []
        for rank in pairs[:2]:
            result.extend([card for card in cards_list if card.get_rank() == rank][:2])
        return result
    return []

def one_pair(cards_list):
    rank_dict = get_rank_dict(cards_list)
    for rank, count in rank_dict.items():
        if count == 2:
            return [card for card in cards_list if card.get_rank() == rank]
    return []

def flush(cards_list):
    suit_dict = get_suit_dict(cards_list)
    for suit, suited_cards in suit_dict.items():
        if len(suited_cards) >= 5:
            return suited_cards[:5]
    return []

def straight(cards_list):
    ranks = sorted(set(card.get_rank() for card in cards_list))
    for i in range(len(ranks) - 4):
        if ranks[i+4] - ranks[i] == 4:
            return [card for card in cards_list if card.get_rank() in range(ranks[i], ranks[i]+5)]
    return []

def straight_flush(cards_list):
    suit_dict = get_suit_dict(cards_list)
    for suit, suited_cards in suit_dict.items():
        if len(suited_cards) >= 5:
            result = straight(suited_cards)
            if result:
                return result
    return []

def determine_category(cards_list):
    if (result := straight_flush(cards_list)):
        return "straight flush", result
    elif (result := four_of_a_kind(cards_list)):
        return "four of a kind", result
    elif (result := full_house(cards_list)):
        return "full house", result
    elif (result := flush(cards_list)):
        return "flush", result
    elif (result := straight(cards_list)):
        return "straight", result
    elif (result := three_of_a_kind(cards_list)):
        return "three of a kind", result
    elif (result := two_pair(cards_list)):
        return "two pair", result
    elif (result := one_pair(cards_list)):
        return "one pair", result
    else:
        return "high card", []

def main():
    print("----------------------------------------")
    print("Let's play poker!")

    deck = cards.Deck()
    deck.shuffle()

    while deck.cards_left() >= 9:
        community = [deck.deal() for _ in range(5)]
        p1 = [deck.deal() for _ in range(2)]
        p2 = [deck.deal() for _ in range(2)]

        print("\nCommunity cards:", community)
        print("Player 1:", p1)
        print("Player 2:", p2)

        hand1 = community + p1
        hand2 = community + p2

        category1, cards1 = determine_category(hand1)
        category2, cards2 = determine_category(hand2)

        categories = [
            "high card", "one pair", "two pair", "three of a kind",
            "straight", "flush", "full house", "four of a kind", "straight flush"
        ]
        val1 = categories.index(category1)
        val2 = categories.index(category2)

        if val1 > val2:
            print(f"\nPlayer 1 wins with {category1}: {cards1}")
        elif val2 > val1:
            print(f"\nPlayer 2 wins with {category2}: {cards2}")
        else:
            print(f"\nTIE with {category1}: {cards1}")

        ans = input("\nDo you wish to play another hand? (Y or N) ").strip().lower()
        if ans != 'y':
            break
        print("----------------------------------------")

    if deck.cards_left() < 9:
        print("Deck has too few cards so game is done.")

if __name__ == "__main__":
    main()
```

```python
import cards

GRID_SIZE = 16
DISCARD_SIZE = 4
TOTAL_SPOTS = GRID_SIZE + DISCARD_SIZE

def create_board():
    return [None] * TOTAL_SPOTS

def print_board(board, current_card):
    print(f"\nDeal:  {current_card}")
    print("Tableau:")
    for row in range(4):
        row_str = ""
        for col in range(5):
            index = row * 5 + col
            if index >= GRID_SIZE:
                break
            cell = board[index]
            row_str += f"{str(cell) if cell else index+1:>5}"
        print(row_str)
    print("\nDiscards: ", end="")
    for i in range(GRID_SIZE, TOTAL_SPOTS):
        print(f"{str(board[i]) if board[i] else i+1}", end="  ")
    print("\n")

def valid_input(user_input, board):
    if not user_input.isdigit():
        return False
    pos = int(user_input)
    if not (1 <= pos <= TOTAL_SPOTS):
        return False
    if board[pos - 1] is not None:
        return False
    return True

def play_game():
    deck = cards.Deck()
    deck.shuffle()
    board = create_board()

    cards_placed = 0

    while cards_placed < GRID_SIZE:
        current_card = deck.deal()
        print_board(board, current_card)

        while True:
            move = input("Choose location (1‚Äì20) to move card to (q to quit): ").strip()
            if move.lower() == 'q':
                print("Game quit early.")
                return
            if valid_input(move, board):
                position = int(move) - 1
                board[position] = current_card
                cards_placed += 1 if position < GRID_SIZE else 0
                break
            else:
                print("Invalid input. Please choose a valid, unoccupied position (1‚Äì20).")

    print_board(board, None)
    print("Game complete. You may now calculate the score (coming soon!).")

if __name__ == "__main__":
    play_game()
```

```python
from cards import Deck

NUM_COLUMNS = 7
NUM_CARDS_PER_COLUMN = 3
NUM_FOUNDATIONS = 4

def initialize_game():
    stock = Deck()
    stock.shuffle()

    tableau = [[] for _ in range(NUM_COLUMNS)]
    foundations = [[] for _ in range(NUM_FOUNDATIONS)]

    for i in range(NUM_CARDS_PER_COLUMN):
        for col in tableau:
            card = stock.deal()
            if i < NUM_CARDS_PER_COLUMN - 1:
                card.set_hidden()
            col.append(card)

    return stock, tableau, foundations

def display_game(stock, tableau, foundations):
    print("\nFOUNDATIONS:")
    for i, f in enumerate(foundations):
        print(f" F{i}: {f[-1] if f else 'Empty'}")
    print("\nTABLEAU:")
    for i, col in enumerate(tableau):
        print(f" T{i}: ", end="")
        for card in col:
            print(str(card), end=" ")
        print()
    print(f"\nStock has {stock.cards_left()} cards left.")

def can_move_to_foundation(card, foundation):
    if not foundation:
        return card.get_rank() == 1
    top = foundation[-1]
    return top.suit == card.suit and card.get_rank() == top.get_rank() + 1

def can_stack(card1, card2):
    return card1.get_rank() + 1 == card2.get_rank() and card1.has_different_color(card2)

def move_to_foundation(tableau, foundations, row, f_index):
    if not tableau[row]:
        print("No cards in that tableau row.")
        return
    card = tableau[row][-1]
    if can_move_to_foundation(card, foundations[f_index]):
        foundations[f_index].append(tableau[row].pop())
        if tableau[row] and tableau[row][-1].is_hidden():
            tableau[row][-1].show_card()
    else:
        print("Illegal move to foundation.")

def move_tableau(tableau, n, src, dest):
    if len(tableau[src]) < n:
        print("Not enough cards to move.")
        return
    moving_cards = tableau[src][-n:]
    if tableau[dest]:
        if not can_stack(moving_cards[0], tableau[dest][-1]):
            print("Invalid move.")
            return
    tableau[dest].extend(moving_cards)
    del tableau[src][-n:]
    if tableau[src] and tableau[src][-1].is_hidden():
        tableau[src][-1].show_card()

def deal_cards(stock, tableau):
    if stock.cards_left() < NUM_COLUMNS:
        print("No cards left in stock.")
        return
    for col in tableau:
        col.append(stock.deal())

def play():
    stock, tableau, foundations = initialize_game()

    print("Welcome to Easthaven Solitaire!")
    display_game(stock, tableau, foundations)

    while True:
        command = input("\nEnter command (h for help): ").strip().lower()

        if command == 'h':
            print("Commands:\n f [row] [foundation] - move to foundation\n m [n] [from] [to] - move cards\n d - deal\n q - quit")
        elif command.startswith('f'):
            parts = command.split()
            if len(parts) == 3:
                move_to_foundation(tableau, foundations, int(parts[1]), int(parts[2]))
            else:
                print("Invalid 'f' command format.")
        elif command.startswith('m'):
            parts = command.split()
            if len(parts) == 4:
                move_tableau(tableau, int(parts[1]), int(parts[2]), int(parts[3]))
            else:
                print("Invalid 'm' command format.")
        elif command == 'd':
            deal_cards(stock, tableau)
        elif command == 'q':
            print("Goodbye!")
            break
        else:
            print("Unknown command.")

        display_game(stock, tableau, foundations)
        if all(len(f) == 13 for f in foundations):
            print("You win!")
            break

if __name__ == "__main__":
    play()
```

```python
from cards import Deck

NUM_COLUMNS = 4

def display_game(tableau, foundation, stock_count):
    print("\nFOUNDATION:")
    for card in foundation:
        print(card, end=" ")
    print(f"\n\nStock cards left: {stock_count}")
    print("\nTABLEAU:")
    for i, pile in enumerate(tableau):
        print(f"Col {i+1}: ", end="")
        if pile:
            for card in pile:
                print(card, end=" ")
        else:
            print("Empty", end="")
        print()

def can_remove(tableau, idx):
    card = tableau[idx][-1]
    for i in range(NUM_COLUMNS):
        if i != idx and tableau[i]:
            other = tableau[i][-1]
            if other.suit == card.suit and other.get_rank() > card.get_rank():
                return True
    return False

def move_card(tableau, src, dest):
    if not tableau[src]:
        print("No card to move in source.")
        return False
    if tableau[dest]:
        print("Destination is not empty.")
        return False
    tableau[dest].append(tableau[src].pop())
    return True

def deal_cards(deck, tableau):
    if any(not row for row in tableau):
        print("Cannot deal: There are empty tableau columns.")
        return False
    if deck.cards_left() < NUM_COLUMNS:
        print("Not enough cards to deal.")
        return False
    for row in tableau:
        row.append(deck.deal())
    return True

def has_won(tableau, stock):
    for row in tableau:
        if any(card.get_rank() != 1 for card in row):
            return False
    return stock.cards_left() == 0

def play():
    deck = Deck()
    deck.shuffle()

    tableau = [[] for _ in range(NUM_COLUMNS)]
    foundation = []

    for row in tableau:
        row.append(deck.deal())

    print("Welcome to Aces Up Solitaire!")
    display_game(tableau, foundation, deck.cards_left())

    while True:
        command = input("\nEnter command (1-4 to remove, m src dest, d to deal, q to quit): ").strip().lower()

        if command == 'q':
            print("Goodbye!")
            break
        elif command == 'd':
            deal_cards(deck, tableau)
        elif command.startswith('m'):
            parts = command.split()
            if len(parts) == 3:
                src = int(parts[1]) - 1
                dest = int(parts[2]) - 1
                if 0 <= src < NUM_COLUMNS and 0 <= dest < NUM_COLUMNS:
                    move_card(tableau, src, dest)
                else:
                    print("Invalid column numbers.")
            else:
                print("Invalid move format. Use: m src dest")
        elif command in ['1', '2', '3', '4']:
            idx = int(command) - 1
            if tableau[idx]:
                if can_remove(tableau, idx):
                    foundation.append(tableau[idx].pop())
                else:
                    print("Cannot remove: No higher card of same suit in another column.")
            else:
                print("No card to remove.")
        else:
            print("Invalid command.")

        display_game(tableau, foundation, deck.cards_left())

        if has_won(tableau, deck):
            print("You win! Only Aces remain.")
            break
        elif deck.cards_left() == 0 and not any(can_remove(tableau, i) for i in range(NUM_COLUMNS)):
            print("No more moves possible. You lose.")
            break

if __name__ == "__main__":
    play()
```

```python
import cards
import random

NUM_STACKS = 10
INITIAL_SCORE = 500

def create_full_deck():
    all_cards = []
    for _ in range(8):
        for rank in range(1, 14):
            all_cards.append(cards.Card(rank, 3))
    random.shuffle(all_cards)
    return all_cards

def deal_initial_tableau(all_cards):
    tableau = [[] for _ in range(NUM_STACKS)]
    for i in range(NUM_STACKS):
        num_cards = 6 if i < 4 else 5
        for _ in range(num_cards):
            card = all_cards.pop()
            tableau[i].append(card)
    return tableau, all_cards

def print_tableau(tableau):
    print("\nCurrent Tableau:")
    max_height = max(len(stack) for stack in tableau)
    for level in range(max_height):
        for stack in tableau:
            if level < len(stack):
                print(str(stack[level]), end="\t")
            else:
                print("  ", end="\t")
        print()
    print("-" * 80)

def move_cards(tableau, n, src, dst):
    try:
        src_stack = tableau[src]
        dst_stack = tableau[dst]
    except IndexError:
        print("Invalid stack index.")
        return False

    if n <= 0 or n > len(src_stack):
        print("Invalid number of cards.")
        return False

    moving_cards = src_stack[-n:]

    if dst_stack:
        top_dst = dst_stack[-1]
        if top_dst.get_rank() != moving_cards[0].get_rank() + 1:
            print("Invalid move. Destination must be 1 rank higher.")
            return False

    dst_stack.extend(moving_cards)
    del src_stack[-n:]
    return True

def deal_from_stock(tableau, stock_cards):
    if len(stock_cards) < NUM_STACKS:
        print("Not enough cards in stock.")
        return False

    if not all(len(stack) > 0 for stack in tableau):
        print("Each stack must have at least one card to deal.")
        return False

    for i in range(NUM_STACKS):
        tableau[i].append(stock_cards.pop())

    return True

def play():
    print("üï∑Ô∏è  Spider Solitaire (Beginner - One Suit)")
    print("Commands:")
    print("  m <n> <src> <dst> : move n cards from stack src to dst")
    print("  d                 : deal a new row from stock")
    print("  h                 : help")
    print("  q                 : quit")

    all_cards = create_full_deck()
    tableau, stock_cards = deal_initial_tableau(all_cards)
    score = INITIAL_SCORE

    while True:
        print_tableau(tableau)
        print(f"Score: {score}")
        command = input("Enter command: ").strip().lower()

        if command == 'q':
            print("Thanks for playing!")
            break
        elif command == 'h':
            print("Commands:")
            print("  m <n> <src> <dst> : move n cards from stack src to dst")
            print("  d                 : deal a new row from stock")
            print("  h                 : help")
            print("  q                 : quit")
        elif command == 'd':
            if deal_from_stock(tableau, stock_cards):
                score -= 1
        elif command.startswith('m'):
            try:
                _, n, src, dst = command.split()
                n, src, dst = int(n), int(src), int(dst)
                if move_cards(tableau, n, src, dst):
                    score -= 1
            except:
                print("Invalid move command. Use: m <n> <src> <dst>")
        else:
            print("Unknown command. Type 'h' for help.")

if __name__ == "__main__":
    play()
```

```python
import itertools

class Matrix:
    def __init__(self):
        self._matrix = []

    def read_file(self, fp):
        num = int(fp.readline())
        self._matrix = [set() for _ in range(num)]
        for line in fp:
            a, b = map(int, line.split())
            self._matrix[a - 1].add(b)
            self._matrix[b - 1].add(a)

    def __str__(self):
        result = ''
        for i, s in enumerate(self._matrix, 1):
            result += f"{i}: {' '.join(str(x) for x in sorted(s))}\n"
        return result

    def adjacent(self, room):
        return self._matrix[room - 1]

    def rooms(self):
        return len(self._matrix)

def open_file():
    while True:
        try:
            filename = input("Enter a file name: ")
            return open(filename, 'r')
        except FileNotFoundError:
            print("File not found, try again.")

def main():
    fp = open_file()
    matrix = Matrix()
    matrix.read_file(fp)
    num_rooms = matrix.rooms()
    all_rooms = set(range(1, num_rooms + 1))

    done = False
    for ta_count in range(1, num_rooms + 1):
        for combo in itertools.combinations(all_rooms, ta_count):
            covered = set()
            for r in combo:
                covered.add(r)
                covered.update(matrix.adjacent(r))
            if covered == all_rooms:
                print(f"TAs needed: {ta_count}")
                print(f"TAs assigned to rooms: {', '.join(map(str, sorted(combo)))}")
                done = True
                break
        if done:
            break

    print("\nAdjacency Matrix")
    print(matrix)

if __name__ == "__main__":
    main()
```

```python
import urllib.request

class Currency:
    def __init__(self, amount=0, code='USD'):
        valid_codes = ['USD', 'EUR', 'SEK', 'CAD', 'CNY', 'GBP']
        if not isinstance(amount, (int, float)) or code not in valid_codes:
            self.amount = 0
            self.code = ''
        else:
            self.amount = float(amount)
            self.code = code

    def __str__(self):
        return f"{self.amount:.2f} {self.code}"

    def __repr__(self):
        return self.__str__()

    def convert_to(self, new_code):
        if self.code == '' or new_code not in ['USD', 'EUR', 'SEK', 'CAD', 'CNY', 'GBP']:
            return Currency(0, '')
        url = f"https://www.google.com/finance/converter?a={self.amount}&from={self.code}&to={new_code}"
        try:
            web_obj = urllib.request.urlopen(url)
            result = str(web_obj.read())
            web_obj.close()

            idx1 = result.find('<span class=bld>')
            idx2 = result.find('</span>', idx1)
            value_str = result[idx1+18:idx2].split()[0]
            new_amount = float(value_str)
            return Currency(new_amount, new_code)
        except:
            return Currency(0, '')

    def __add__(self, other):
        if isinstance(other, Currency):
            other_converted = other.convert_to(self.code)
            return Currency(self.amount + other_converted.amount, self.code)
        elif isinstance(other, (int, float)):
            return Currency(self.amount + other, self.code)

    def __radd__(self, other):
        return self.__add__(other)

    def __sub__(self, other):
        if isinstance(other, Currency):
            other_converted = other.convert_to(self.code)
            return Currency(self.amount - other_converted.amount, self.code)
        elif isinstance(other, (int, float)):
            return Currency(self.amount - other, self.code)

    def __rsub__(self, other):
        if isinstance(other, (int, float)):
            return Currency(other - self.amount, self.code)

    def __gt__(self, other):
        if isinstance(other, Currency):
            other_converted = other.convert_to(self.code)
            return self.amount > other_converted.amount
        return False

def run_tests():
    print("=== Running Tests ===")
    c1 = Currency(7.5, 'USD')
    c2 = Currency(2, 'EUR')
    print("Original currencies:", c1, c2)

    print("Conversion of c2 to USD:")
    c3 = c2.convert_to('USD')
    print(c3)

    print("Addition with Currency and float:")
    print("c1 + c2 =", c1 + c2)
    print("c1 + 5.5 =", c1 + 5.5)
    print("5 + c1 =", 5 + c1)

    print("Subtraction with Currency and float:")
    print("c1 - c2 =", c1 - c2)
    print("c1 - 3 =", c1 - 3)
    print("10 - c1 =", 10 - c1)

    print("Greater than comparison:")
    print("c1 > c2 =", c1 > c2)
    print("======================\n")

def run_app():
    print("Welcome to the International Expense Tracker.")
    print("Your account starts with 1000 USD.")
    account = Currency(1000, 'USD')

    while True:
        entry = input("Enter an expense (amount CODE) or 'q' to quit: ").strip()
        if entry.lower() == 'q':
            break
        try:
            amount_str, code = entry.split()
            amount = float(amount_str)
            expense = Currency(amount, code.upper())
            if expense.code == '':
                raise ValueError("Invalid currency code.")
            account = account - expense
            print(f"New balance: {account}")
            if Currency(0, 'USD') > account:
                print("Account overdrawn.")
                break
        except ValueError:
            print("Invalid input. Use format: amount CODE (e.g., 25 USD)")
        except:
            print("An unexpected error occurred.")

if __name__ == "__main__":
    run_tests()
    run_app()
```

```python
import urllib.request
import json

class Tour:
    def __init__(self, *cities):
        self.cities = list(cities)

    def __str__(self):
        return "; ".join(self.cities)

    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        return Tour(*(self.cities + other.cities))

    def __mul__(self, times):
        if not isinstance(times, int):
            raise TypeError("Multiplication only supports non-negative integers")
        if times < 0:
            raise ValueError("Cannot multiply Tour by a negative number")
        return Tour(*(self.cities * times))

    def __rmul__(self, times):
        return self.__mul__(times)

    def __eq__(self, other):
        return self.cities == other.cities

    def __gt__(self, other):
        return self.distance() > other.distance()

    def __lt__(self, other):
        return self.distance() < other.distance()

    def distance(self, mode="driving"):
        total_distance = 0
        for i in range(len(self.cities) - 1):
            origin = self.cities[i].replace(" ", "+")
            destination = self.cities[i+1].replace(" ", "+")
            url = f"http://maps.googleapis.com/maps/api/distancematrix/json?origins={origin}&destinations={destination}&mode={mode}&sensor=false"

            try:
                web_obj = urllib.request.urlopen(url)
                results_str = web_obj.read().decode("utf-8")
                web_obj.close()
                data = json.loads(results_str)

                if data["rows"][0]["elements"][0]["status"] != "OK":
                    raise ValueError(f"No valid route between {self.cities[i]} and {self.cities[i+1]} using mode '{mode}'.")

                distance_value = data["rows"][0]["elements"][0]["distance"]["value"]
                total_distance += distance_value

            except Exception as e:
                raise ValueError(f"Error calculating distance between {self.cities[i]} and {self.cities[i+1]}: {e}")

        return total_distance


def main():
    t1 = Tour("New York, NY", "Lansing, MI", "Sacramento, CA")
    t2 = Tour("Oakland, CA")
    t3 = Tour("Sacramento, CA", "Oakland, CA")

    print("t1:", t1)
    print("t2:", t2)
    print("t3:", t3)

    try:
        print("t1 distances: driving-{} km; biking-{} km; walking-{} km".format(
            round(t1.distance()/1000),
            round(t1.distance('bicycling')/1000),
            round(t1.distance('walking')/1000)
        ))
    except ValueError as e:
        print("Error getting distances for t1:", e)

    print("\nUsing driving distances from here on.")
    t4 = t1 + t2
    print("t4:", t4)
    try:
        print("t4 driving distance:", round(t4.distance()/1000), "km")
    except ValueError as e:
        print("Error getting t4 distance:", e)

    print("t4 == t1 + t2:", t4 == (t1 + t2))

main()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class SnowPerson:
    def __init__(self, ax, x, y):
        self.ax = ax
        self.x = x
        self.y = y
        self.draw_body()
        self.draw_face()

    def draw_circle(self, x, y, radius, edge='black', fill='white'):
        circle = patches.Circle((x, y), radius, edgecolor=edge, facecolor=fill, linewidth=2)
        self.ax.add_patch(circle)

    def draw_line(self, x1, y1, x2, y2, color='black'):
        line = plt.Line2D([x1, x2], [y1, y2], color=color, linewidth=2)
        self.ax.add_line(line)

    def draw_rect(self, x, y, width, height, edge='black', fill='black'):
        rect = patches.Rectangle((x, y), width, height, edgecolor=edge, facecolor=fill, linewidth=2)
        self.ax.add_patch(rect)

    def draw_triangle(self, x, y, size, edge='black', fill='black'):
        triangle = patches.Polygon([[x, y], [x + size / 2, y + size], [x + size, y]],
                                   closed=True, edgecolor=edge, facecolor=fill, linewidth=2)
        self.ax.add_patch(triangle)

    def draw_body(self):
        self.draw_circle(self.x, self.y, 30)
        self.draw_circle(self.x, self.y + 60, 20)
        self.draw_circle(self.x, self.y + 100, 15)

    def draw_face(self):
        self.draw_circle(self.x - 5, self.y + 110, 2, fill='black')
        self.draw_circle(self.x + 5, self.y + 110, 2, fill='black')
        self.draw_line(self.x - 5, self.y + 100, self.x + 5, self.y + 100)

class SnowMan(SnowPerson):
    def draw_extras(self):
        self.draw_line(self.x - 20, self.y + 60, self.x - 50, self.y + 80, 'brown')
        self.draw_line(self.x + 20, self.y + 60, self.x + 50, self.y + 80, 'brown')
        self.draw_rect(self.x - 15, self.y + 120, 30, 5)
        self.draw_rect(self.x - 10, self.y + 125, 20, 20)

    def __init__(self, ax, x, y):
        super().__init__(ax, x, y)
        self.draw_extras()

class SnowLady(SnowPerson):
    def draw_extras(self):
        self.draw_rect(self.x - 10, self.y + 90, 20, 5, fill='red')
        self.draw_triangle(self.x - 10, self.y + 120, 20, fill='purple')

    def __init__(self, ax, x, y):
        super().__init__(ax, x, y)
        self.draw_extras()

def main():
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim(-150, 150)
    ax.set_ylim(-50, 200)
    ax.axis('off')

    SnowMan(ax, -60, 0)
    SnowLady(ax, 60, 0)

    plt.show()

main()
```

```python
from google.colab import files
uploaded = files.upload()

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math

class Rectangle:
    def __init__(self, x, y, width, height, color):
        self.x = int(x)
        self.y = int(y)
        self.width = int(width)
        self.height = int(height)
        self.color = color.strip()

    def draw(self, ax):
        bottom_left = (self.x - self.width / 2, self.y - self.height / 2)
        rect = patches.Rectangle(bottom_left, self.width, self.height,
                                 linewidth=1, edgecolor='black', facecolor=self.color)
        ax.add_patch(rect)

    def get_bounds(self):
        return (self.x - self.width / 2, self.y - self.height / 2,
                self.x + self.width / 2, self.y + self.height / 2)

    def __str__(self):
        return f"Rectangle x:{self.x}, y:{self.y}, width:{self.width}, height:{self.height}, color:{self.color}"

class Star:
    def __init__(self, x, y, arm_length, color):
        self.x = int(x)
        self.y = int(y)
        self.arm = int(arm_length)
        self.color = color.strip()

    def draw(self, ax):
        points = []
        for i in range(10):
            angle = math.radians(i * 36 - 90)
            radius = self.arm if i % 2 == 0 else self.arm / 2
            px = self.x + radius * math.cos(angle)
            py = self.y + radius * math.sin(angle)
            points.append((px, py))
        star = patches.Polygon(points, closed=True, edgecolor='black', facecolor=self.color)
        ax.add_patch(star)

    def get_bounds(self):
        return (self.x - self.arm, self.y - self.arm, self.x + self.arm, self.y + self.arm)

    def __str__(self):
        return f"Star x:{self.x}, y:{self.y}, arm:{self.arm}, color:{self.color}"

class Flag:
    def __init__(self, file_path):
        self.rectangles = []
        self.stars = []
        with open(file_path, 'r') as f:
            num_rects = int(f.readline())
            for _ in range(num_rects):
                parts = f.readline().strip().split(',')
                self.rectangles.append(Rectangle(*parts))
            num_stars = int(f.readline())
            for _ in range(num_stars):
                parts = f.readline().strip().split(',')
                self.stars.append(Star(*parts))

    def draw(self):
        fig, ax = plt.subplots(figsize=(8, 6))
        ax.set_aspect('equal')
        ax.axis('off')
        min_x, min_y, max_x, max_y = 0, 0, 0, 0
        for r in self.rectangles:
            x1, y1, x2, y2 = r.get_bounds()
            min_x, min_y = min(min_x, x1), min(min_y, y1)
            max_x, max_y = max(max_x, x2), max(max_y, y2)
        for s in self.stars:
            x1, y1, x2, y2 = s.get_bounds()
            min_x, min_y = min(min_x, x1), min(min_y, y1)
            max_x, max_y = max(max_x, x2), max(max_y, y2)

        padding = 50
        ax.set_xlim(min_x - padding, max_x + padding)
        ax.set_ylim(min_y - padding, max_y + padding)
        for rect in self.rectangles:
            rect.draw(ax)
        for star in self.stars:
            star.draw(ax)

        plt.show()

    def __str__(self):
        output = "Rectangles\n"
        for r in self.rectangles:
            output += str(r) + "\n"
        output += "Stars\n"
        for s in self.stars:
            output += str(s) + "\n"
        return output

flag_file = 'senegal.txt'
flag = Flag(flag_file)
print(flag)
flag.draw()
```

```python
from google.colab import files
uploaded = files.upload()

import math

class Quad:
    def __init__(self, ab, da=None, angleA=90):
        self.valid = False
        try:
            ab = float(ab)
            if da is None:
                da = ab
            else:
                da = float(da)
            angleA = float(angleA)
            if ab > 0 and da > 0 and 0 < angleA < 360:
                self.ab = ab
                self.da = da
                self.angleA = angleA
                self.valid = True
            else:
                self.ab = self.da = self.angleA = 0
        except:
            self.ab = self.da = self.angleA = 0

    def __str__(self):
        if self.valid:
            return f"[ {self.ab:.1f}, {self.da:.1f}, {self.ab:.1f}, {self.da:.1f} ]"
        else:
            return "[ invalid ]"

    def sides(self):
        if self.valid:
            return (self.ab, self.da, self.ab, self.da)
        else:
            return (0, 0, 0, 0)

    def angles(self):
        if self.valid:
            A = self.angleA
            B = 180 - A
            return (A, B, A, B)
        else:
            return (None, None, None, None)

    def perimeter(self):
        if self.valid:
            return 2 * (self.ab + self.da)
        else:
            return 0

    def area(self):
        if self.valid:
            radians = math.radians(self.angleA)
            return self.ab * self.da * math.sin(radians)
        else:
            return 0

    def scale(self, factor):
        try:
            factor = float(factor)
            if self.valid and factor > 0:
                self.ab *= factor
                self.da *= factor
        except:
            pass

file_name = "quad_input.txt"

try:
    with open(file_name, "r") as file:
        line_count = 0
        valid_quads = []
        total_perimeter = 0
        total_area = 0

        for line in file:
            line_count += 1
            line = line.strip()
            print(f"Line {line_count}: {line}")
            if not line:
                print("No input data\n")
                continue

            parts = line.split(',')
            try:
                ab = float(parts[0])
                da = float(parts[1]) if len(parts) > 1 else None
                angle = float(parts[2]) if len(parts) > 2 else 90
                quad = Quad(ab, da, angle)
            except:
                print("Invalid input format\n")
                continue

            if quad.valid:
                print(f"{quad}")
                peri = quad.perimeter()
                area = quad.area()
                print(f"Perimeter = {peri:.1f}")
                print(f"Area = {area:.1f}")
                valid_quads.append((quad, peri, area))
                total_perimeter += peri
                total_area += area
            else:
                print("Invalid parallelogram")
            print()

        print("Summary:")
        print(f"Total lines processed: {line_count}")
        print(f"Valid parallelograms: {len(valid_quads)}")

        if valid_quads:
            avg_perimeter = total_perimeter / len(valid_quads)
            avg_area = total_area / len(valid_quads)
            print(f"Average perimeter: {avg_perimeter:.1f}")
            print(f"Average area: {avg_area:.1f}")

            max_peri_quad = max(valid_quads, key=lambda x: x[1])
            max_area_quad = max(valid_quads, key=lambda x: x[2])

            print("\nParallelogram with largest perimeter:")
            print(f"{max_peri_quad[0]}, Perimeter = {max_peri_quad[1]:.1f}, Area = {max_peri_quad[2]:.1f}")

            print("\nParallelogram with largest area:")
            print(f"{max_area_quad[0]}, Perimeter = {max_area_quad[1]:.1f}, Area = {max_area_quad[2]:.1f}")

except FileNotFoundError:
    print("Error: Please upload a file named 'quad_input.txt'")
```

```python
import math

class Triangle:
    def __init__(self, a, b, c):
        try:
            self.a = float(a)
            self.b = float(b)
            self.c = float(c)
        except:
            self.a = self.b = self.c = 0
        self.valid = self._is_valid()

    def _is_valid(self):
        a, b, c = self.a, self.b, self.c
        return all(x > 0 for x in (a, b, c)) and (a + b > c and a + c > b and b + c > a)

    def __str__(self):
        return f"( {self.a:.1f}, {self.b:.1f}, {self.c:.1f} )" if self.valid else "( invalid )"

    def sides(self):
        return (self.a, self.b, self.c)

    def angles(self):
        if not self.valid:
            return (None, None, None)
        a, b, c = self.a, self.b, self.c
        A = math.degrees(math.acos((b**2 + c**2 - a**2)/(2*b*c)))
        B = math.degrees(math.acos((a**2 + c**2 - b**2)/(2*a*c)))
        C = 180 - A - B
        return (round(A, 1), round(B, 1), round(C, 1))

    def perimeter(self):
        return self.a + self.b + self.c if self.valid else 0

    def area(self):
        if not self.valid:
            return 0
        s = self.perimeter() / 2
        T = s * (s - self.a) * (s - self.b) * (s - self.c)
        return math.sqrt(T) if T > 0 else 0

    def scale(self, factor):
        try:
            factor = float(factor)
            if factor > 0 and self.valid:
                self.a *= factor
                self.b *= factor
                self.c *= factor
        except:
            pass

triangle_data = [
    "( 3, 4, 5 )",
    "( 5, 5, 5 )",
    "( 1, 2, 3 )",
    "( 6, 8, 10 )",
    "( 0, 5, 5 )",
    "( -1, 4, 5 )",
    "( 7, 24, 25 )"
]

triangles = []
print("Processing triangles:\n")
for idx, line in enumerate(triangle_data, start=1):
    print(f"Line {idx}: {line}")
    try:
        values = line.strip("() ").split(',')
        t = Triangle(*values)
        if t.valid:
            peri = t.perimeter()
            area = t.area()
            print(f"Triangle: {t}")
            print(f"Perimeter: {peri:.1f}")
            print(f"Area: {area:.1f}")
            triangles.append((t, peri, area))
        else:
            print("Invalid triangle")
    except:
        print("Error processing line")
    print()

if triangles:
    total = len(triangle_data)
    valid = len(triangles)
    avg_peri = sum(t[1] for t in triangles) / valid
    avg_area = sum(t[2] for t in triangles) / valid
    max_peri = max(triangles, key=lambda x: x[1])
    max_area = max(triangles, key=lambda x: x[2])

    print("Summary:")
    print(f"Total lines processed: {total}")
    print(f"Valid triangles: {valid}")
    print(f"Average perimeter: {avg_peri:.1f}")
    print(f"Average area: {avg_area:.1f}")
    print(f"Largest perimeter: {max_peri[0]} (Perimeter: {max_peri[1]:.1f}, Area: {max_peri[2]:.1f})")
    print(f"Largest area: {max_area[0]} (Perimeter: {max_area[1]:.1f}, Area: {max_area[2]:.1f})")
else:
    print("No valid triangles found.")
```

```python
import random

class Customer:
    def __init__(self, cid, num_floors):
        self.ID = cid
        self.cur_floor = random.randint(1, num_floors)
        self.dst_floor = random.randint(1, num_floors)
        while self.dst_floor == self.cur_floor:
            self.dst_floor = random.randint(1, num_floors)
        self.in_elevator = False
        self.finished = False

    def __str__(self):
        return f"Customer {self.ID} from {self.cur_floor} to {self.dst_floor}"

class Elevator:
    def __init__(self, num_floors):
        self.num_floors = num_floors
        self.cur_floor = 1
        self.direction = 1
        self.register_list = []

    def move(self):
        self.cur_floor += self.direction
        if self.cur_floor == self.num_floors or self.cur_floor == 1:
            self.direction *= -1

    def register_customer(self, customer):
        self.register_list.append(customer)
        customer.in_elevator = True

    def cancel_customer(self, customer):
        self.register_list.remove(customer)
        customer.in_elevator = False
        customer.finished = True

class Building:
    def __init__(self, num_floors, num_customers):
        self.num_floors = num_floors
        self.customers = [Customer(i+1, num_floors) for i in range(num_customers)]
        self.elevator = Elevator(num_floors)
        self.total_moves = 0

    def run(self):
        print("\nStarting Elevator Simulation...\n")
        while not all(c.finished for c in self.customers):
            self.output()
            for customer in self.elevator.register_list[:]:
                if customer.dst_floor == self.elevator.cur_floor:
                    self.elevator.cancel_customer(customer)

            for customer in self.customers:
                if not customer.in_elevator and not customer.finished and customer.cur_floor == self.elevator.cur_floor:
                    self.elevator.register_customer(customer)

            self.elevator.move()
            self.total_moves += 1

        print(f"\nSimulation Complete! Total elevator moves: {self.total_moves}\n")

    def output(self):
        print(f"Elevator at floor {self.elevator.cur_floor} | Direction: {'Up' if self.elevator.direction == 1 else 'Down'}")
        print("In elevator:", [c.ID for c in self.elevator.register_list])
        print("Waiting:", [c.ID for c in self.customers if not c.finished and not c.in_elevator])
        print()

def main():
    while True:
        try:
            floors = int(input("Enter number of floors: "))
            customers = int(input("Enter number of customers: "))
            if floors < 2 or customers < 1:
                print("Please enter at least 2 floors and 1 customer.")
                continue
            break
        except ValueError:
            print("Please enter valid integers.")

    b = Building(floors, customers)
    b.run()

main()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class Circle:
    def __init__(self, color='black', fill=True):
        self.color = color
        self.fill = fill

    def draw(self, ax, cx, cy, r):
        circle = patches.Circle((cx, cy), r, edgecolor=self.color,
                                facecolor=self.color if self.fill else 'none')
        ax.add_patch(circle)

class Rectangle:
    def __init__(self, color='black', fill=True):
        self.color = color
        self.fill = fill

    def draw(self, ax, x, y, width, height):
        rect = patches.Rectangle((x, y), width, height, edgecolor=self.color,
                                 facecolor=self.color if self.fill else 'none')
        ax.add_patch(rect)

class Eye:
    def __init__(self, eye_color='white', pupil_color='black'):
        self.eye = Circle(eye_color, True)
        self.pupil = Circle(pupil_color, True)

    def draw(self, ax, cx, cy):
        self.eye.draw(ax, cx, cy, 10)
        self.pupil.draw(ax, cx, cy, 4)

class Smile:
    def __init__(self, color='black'):
        self.color = color

    def draw(self, ax, x, y, width, height):
        smile = patches.Arc((x, y), width, height, angle=0, theta1=200, theta2=340,
                            edgecolor=self.color, linewidth=2)
        ax.add_patch(smile)

class Hair:
    def __init__(self, color='black'):
        self.color = color

    def draw(self, ax):
        for i in range(5):
            x = 120 + i * 20
            Rectangle(self.color).draw(ax, x, 230, 10, 20)

def face():
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim(100, 300)
    ax.set_ylim(100, 300)
    ax.axis('off')

    head = Circle('orange')
    head.draw(ax, 200, 200, 80)
    left_eye = Eye()
    right_eye = Eye()
    left_eye.draw(ax, 170, 220)
    right_eye.draw(ax, 230, 220)
    smile = Smile()
    smile.draw(ax, 200, 180, 60, 30)
    hair = Hair()
    hair.draw(ax)

    plt.show()

face()
```
